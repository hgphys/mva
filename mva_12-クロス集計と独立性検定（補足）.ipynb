## **1. 量的データと質的データの分析の使い分け**

分散分析（ANOVA）が「グループ（質的変数）」による「数値（量的変数）」の差を分析したのに対し、**独立性検定**は「質的変数 × 質的変数」の関係（関連）を分析します。

| 分析手法 | 独立変数（原因側） | 従属変数（結果側） | 例 |
| :--- | :--- | :--- | :--- |
| **分散分析** | 質的（グループ） | **量的（点数など）** | 指導法で平均点は変わるか？ |
| **独立性検定** | 質的（グループ） | **質的（合否など）** | 指導法で合格率は変わるか？ |


## **2. 具体例：指導法と合否のクロス集計表**

3つの指導法（クラスA, B, C）において、合格・不合格の人数が以下のように集計されたとします。

| クラス | 合格 (Pass) | 不合格 (Fail) | 合計 (周辺和) |
| :--- | :---: | :---: | :---: |
| **クラスA** | 40 | 10 | **50** |
| **クラスB** | 30 | 20 | **50** |
| **クラスC** | 15 | 35 | **50** |
| **合計 (周辺和)** | **85** | **65** | **150 ($n$)** |

## **3. 行列としてのクロス集計表と「1ベクトル」**

クロス集計表は、数理的には $r$ 行 $c$ 列の行列 $O$ として表現されます。

$$O = \begin{pmatrix}
40 & 10 \\
30 & 20 \\
15 & 35
\end{pmatrix} \in \mathbb{R}^{3 \times 2}$$

分散分析で全体平均を $\mathbf{1}$ ベクトルとの内積で表現したように、クロス集計の**周辺和（合計値）**も、各次元の $\mathbf{1}$ ベクトルを用いた行列演算として定義できます。

- **行の周辺和ベクトル** $\mathbf{r}$ (各クラスの人数):
行列 $O$ に右から $\mathbf{1}_c$ を掛けて得られます。

$$\mathbf{r} = O \mathbf{1}_c = \begin{pmatrix} 40 & 10 \\ 30 & 20 \\ 15 & 35 \end{pmatrix} \begin{pmatrix} 1 \\ 1 \end{pmatrix} = \begin{pmatrix} 50 \\ 50 \\ 50 \end{pmatrix}$$

- **列の周辺和ベクトル** $\mathbf{c}$ (全体の合格・不合格数):
行列 $O$ に左から $\mathbf{1}_r^\top$ を掛けて得られます。

$$\mathbf{c}^\top = \mathbf{1}_r^\top O = \begin{pmatrix} 1 & 1 & 1 \end{pmatrix} \begin{pmatrix} 40 & 10 \\ 30 & 20 \\ 15 & 35 \end{pmatrix} = \begin{pmatrix} 85 & 65 \end{pmatrix}$$

- **全サンプルサイズ** $n$:

$$n = \mathbf{1}_r^\top O \mathbf{1}_c = 150$$

## **4. 期待度数：独立性から生まれる「ランク1行列」**

「指導法と合否が完全に無関係（独立）」であるという帰無仮説 $H_0$ の下では、期待度数行列 $E$ は、周辺和の情報の組み合わせだけで決まります。

$$E = \frac{1}{n} \mathbf{r} \mathbf{c}^\top = \frac{1}{150} \begin{pmatrix} 50 \\ 50 \\ 50 \end{pmatrix} \begin{pmatrix} 85 & 65 \end{pmatrix} = \begin{pmatrix}
28.3 & 21.7 \\
28.3 & 21.7 \\
28.3 & 21.7
\end{pmatrix}$$

この $E$ は、全ての行が同じ比率（全体合格率 $85/150$）の定数倍になっているため、行列のランクが 1 （ランク1行列）となります。
つまり、「独立」とは「行列が周辺和の情報（枠組み）だけに依存し、内部に独自のパターンを持たない」状態を指します。

## **5. 独立性検定のプロセス：なぜ検定が必要か？**

現実のデータ行列 $O$ は、期待行列 $E$ と完全には一致しません。そこで、**「観測されたズレが、単なる偶然（ノイズ）なのか、それとも意味のある関連（シグナル）なのか」** を判定するのが**独立性検定（カイ二乗検定）** です。

**① カイ二乗値の計算（ズレの測定）**

$$\chi^2 = \sum_{i,j} \frac{(o_{ij} - e_{ij})^2}{e_{ij}}$$

これは、観測された行列 $O$ がランク1行列 $E$ からどれだけ「突き出しているか」を測る指標です。

**② 自由度の役割（物差しの決定）**

ここで自由度が登場します。自由度は**「ノイズが暴れまわることのできる空間の広さ」** です。

- **自由度が小さい場合**（例：$2 \times 2$ 表、$df=1$）:
ノイズが動ける方向が1つしかないため、偶然大きなズレが生じる確率は低くなります。
- **自由度が大きい場合**（例：$10 \times 10$ 表、$df=81$）:
ノイズが多くのマス目に分散して暴れることができるため、合計としてのズレ（$\chi^2$値）は自然と大きくなりやすくなります。

これら自由度（空間の広さ）に応じた「物差し」と比較して判定を行います。


## **6. 自由度の幾何学的意味**
クロス集計表の自由度が $(r-1)(c-1)$ となる理由は、周辺和という**線形制約**にあります。
|  | 合格 | 不合格 | 合計 |
| :--- | :---: | :---: | :---: |
| **クラスA** | ★ | (自動) | 50 |
| **クラスB** | ★ | (自動) | 50 |
| **クラスC** | (自動) | (自動) | 50 |
| **合計** | 85 | 65 | 150 |

「★」がついた成分（今回の例では2マス）さえ決まれば、残りのマス目は周辺和に合わせるために自動的に決定されます。この **「独立に動かせる変数の数」** が、検定における物差し（カイ二乗分布）の形状を決定します。この「独立に動かせる変数の数」が、カイ二乗検定という判定の舞台において、**「ノイズという霧が広がる次元数」** になります。


## **7. まとめ：分散分析との共通点**

| 概念 | 分散分析 (ANOVA) | 独立性検定 (カイ二乗) |
| :--- | :--- | :--- |
| **データ構造** | $\mathbf{x} \in \mathbb{R}^n$ （ベクトル） | $O \in \mathbb{R}^{r \times c}$ （行列） |
| **合計・平均の計算** | $\mathbf{x} \cdot \mathbf{1}$ （内積） | $O \mathbf{1}_c, \mathbf{1}_r^\top O$ （行列積） |
| **帰無仮説 ($H_0$)** | ベクトルが $\mathbf{1}$ 方向にのみ存在 | 行列が「ランク 1」である |
| **シグナルの正体** | 全体平均からの「射影のズレ」 | 独立性行列からの「成分のズレ」 |
| **自由度の意味** | シグナル空間の次元 | ズレが暴れまわる空間の次元 |

どちらの手法も、 **「低次元のシンプルなモデル（差がない世界）から、現実のデータがどれだけ複雑に突き出しているか」** を検定しているという点で、数理的な構造は共通しています。

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 観測データの行列形式
# クラスごとの合格・不合格者数を並べたクロス集計表を行列として定義します
obs = np.array([
    [40, 10], # クラスA
    [30, 20], # クラスB
    [15, 35]  # クラスC
])

# 2. カイ二乗検定（独立性の検定）の実行
# chi2: カイ二乗値（ズレの合計）
# p: P値（偶然そのズレが生じる確率）
# dof: 自由度（(行数-1)*(列数-1)）
# expected: 期待度数（もし独立だったら理論上こうなるはず、という行列）
chi2, p, dof, expected = stats.chi2_contingency(obs)

print(f"カイ二乗値: {chi2:.4f}")
print(f"自由度: {dof} (=(3-1)*(2-1))")
print(f"P値: {p:.10e}")

# 3. 理論上の期待度数行列の表示
# ランク1行列（すべての行が同じ比率になっている行列）であることを確認できます
print("\n--- 理論上の期待度数行列 (ランク1行列) ---")
print(pd.DataFrame(expected, columns=['Pass', 'Fail'], index=['A', 'B', 'C']))

# 可視化：積み上げ棒グラフで割合を比較
cross_tab = pd.DataFrame(obs, index=['Class_A', 'Class_B', 'Class_C'], columns=['Pass', 'Fail'])
cross_tab_pct = cross_tab.div(cross_tab.sum(axis=1), axis=0)

cross_tab_pct.plot(kind='bar', stacked=True, figsize=(8, 5), color=['#66b3ff','#ff9999'])
plt.title('Pass/Fail Ratio by Class')
plt.ylabel('Ratio')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
