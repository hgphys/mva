## 要点まとめ

**Part VI：情報の最大化**

「データが持つ情報（分散）そのもの」に注目します。データのグループ間に「意味のある差（情報）」があるかどうかを判断するための手法、分散分析（Analysis of Variance: ANOVA） を学びます。分散分析の目的関数である F値 は、「ノイズに対するシグナルの比率（S/N比）」です。この回を通じて、「情報は足し算で分解できる」という感覚を養いましょう。

### **1. 分散分析（ANOVA）とは**

分散分析は、3つ以上のグループの間で **「平均」** に差があるかどうかを、**「分散」** を用いて検定する統計手法です。

| ID | グループ | グループ内ID | データ |
| :---: | :---: | :---: | :---: |
| 1 | 1 | 1 | $x_{11}$ |
| $\vdots$ | $\vdots$ | $\vdots$ | $\vdots$ |
| $m$ | $i$ | $j$ | $x_{ij}$ |
| $\vdots$ | $\vdots$ | $\vdots$ | $\vdots$ |
| $n$ | $k$ | $n_k$ | $x_{kn_k}$ |


例えば、$m$行目のデータにグループ$i$の$j$番目のデータ $x_{ij}$ が格納されているとき、各グループの平均値 $\bar{\mathbf{x}}=(\bar{x}_1, \cdots, \bar{x}_k)^{\mathrm{T}}$ の間に差があるかを統計的に判別するタスクになります。

「グループごとの平均値のばらつき」が、「グループ内の個々人のばらつき（誤差）」よりも十分に大きければ、それは「グループ間に明確な差がある」とみなせる、というロジックに基づいているため、**「平均」** の差を見るのに **「分散」** 分析という名称になっています。

<br>

### **2. シグナルとノイズの分離**

データが持つ「ばらつき（全変動）」は、意味のある「要因によるばらつき（シグナル）」と、無意味な「偶然の誤差（ノイズ）」の和に分解できます。
分散分析は、この比率（F値）を計算し、シグナルがノイズに対して十分に大きいかを検定します。すなわち、分散分析は **「情報のシグナル比（SNR）の評価」** として捉えることができます。

| 概念 | 統計用語 | 数理的意味 | アナロジー |
 | ----- | ----- | ----- | ----- |
| **全情報** | 全変動 ($S_T$) | 全データのばらつきの総和 | 部屋全体の音量 |
| **シグナル** | 級間変動 ($S_A$) | グループ間の平均値の差 | 聞きたい人の会話の声 |
| **ノイズ** | 級内変動 ($S_E$) | グループ内での個体差（誤差） | 周囲の雑音 |
| **判定指標** | F値 | $F = \frac{\text{シグナル}}{\text{ノイズ}}$ | 声が雑音よりどれだけ大きいか (SNR) |

<br>

### **3. 変動の分解とF値**

#### **(1) ベクトルの分解（平方和の分解）**

私たちが分析したい「データのばらつき」は、ベクトル $\mathbf{x}$ を以下の3つのベクトル（全体平均、グループ平均、元データ）を用いて分解することで表現できます。

$$\underbrace{ (\mathbf{x} - \bar{\bar{\mathbf{x}}}) }_{\mathbf{t}: \text{全変動ベクトル}} \quad = (\underbrace{ \bar{\mathbf{x}}_A - \bar{\bar{\mathbf{x}}}}_{\mathbf{a}: \text{シグナルベクトル}})\quad  + \quad (\underbrace{ \mathbf{x} - \bar{\mathbf{x}}_A }_{\mathbf{e}: \text{ノイズベクトル}})$$

数理最適化の視点（シグナル/ノイズ）や統計学の用語とは、以下のように対応します。
また、これらのベクトルの「長さの2乗（2乗ノルム）」が、分散分析で計算する「変動（平方和）」になります。

- $\mathbf{t}$ (Total): 全体平均からのズレ（全情報のばらつき）。統計学的には **全偏差ベクトル** 。ノルムの2乗が **全変動**  ($S_T$) に対応。
$$S_T = \|\mathbf{t}\|^2 = \sum_{i=1}^k \sum_{j=1}^{n_i} (x_{ij} - \bar{\bar{x}})^2$$
- $\mathbf{a}$ (Among): グループ平均が全体平均からどれだけズレているか（シグナル）。統計学的には **要因効果ベクトル** 。ノルムの2乗が **級間変動**  ($S_A$)  に対応。
$$S_A = \|\mathbf{a}\|^2 = \sum_{i=1}^k n_i (\bar{x}_i - \bar{\bar{x}})^2$$
- $\mathbf{e}$ (Error): 個々のデータがグループ平均からどれだけズレているか（ノイズ）。統計学的には **残差ベクトル** 。ノルムの2乗が **級内変動**  ($S_E$) に対応。
$$S_E = \|\mathbf{e}\|^2 = \sum_{i=1}^k \sum_{j=1}^{n_i} (x_{ij} - \bar{x}_i)^2$$

これら2つのベクトル（シグナル $\mathbf{a}$ と ノイズ $\mathbf{e}$）は互いに直交しているため、ピタゴラスの定理により、変動の足し算が成り立ちます。

$$\|\mathbf{t}\|^2 = \|\mathbf{a}\|^2 + \|\mathbf{e}\|^2 \quad \Longleftrightarrow \quad S_T = S_A + S_E$$


#### **(2) 平均平方（分散）への変換とF値**

平方和（$S$）はデータの数に依存するため、自由度（$dof$）で割って「平均平方（$V$）」、つまり分散として比較可能にします。「自由度」とは、それぞれのベクトルが動くことのできる空間の次元のことです。

$$V_A = \frac{S_A}{k-1}=\frac{\|\mathbf{a}\|^2}{\dim(\mathbf{a})}, \quad V_E = \frac{S_E}{n-k}=\frac{\|\mathbf{e}\|^2}{\dim(\mathbf{e})}$$

これらを用いて、最終的な評価指標であるF値は以下のように定義されます。

$$F = \frac{V_A}{V_E} = \frac{\|\mathbf{a}\|^2 / \dim(\mathbf{a})}{\|\mathbf{e}\|^2 / \dim(\mathbf{e})}$$

F値は、シグナル空間とノイズ空間における 「1次元あたりのエネルギー密度（温度）」を比較していることに相当します。この $F$ が、F分布という確率分布に従うことを利用して、「このシグナル比は偶然生じる範囲内か？」を判断します。

## **Python 演習課題**

### **mva_12-A**
✅ **目的**: 幾何学的な「ベクトルの分解（直交性）」と、統計学的な「平方和の分解（総和計算）」が数学的に等価であることを確認しましょう。また、統計学における「自由度」と「ベクトルの次元」との対応について、物理学的な視点「自由に動ける次元」として理解できることを確認しましょう。

#### 🖊 **【数理理解】平方和の分解と自由度**

#### **平方和の分解**
シグナルベクトル $\mathbf{a}$ とノイズベクトル $\mathbf{e}$ の内積が 0 になる（直交する）ことは、総和記号 $\sum$ を用いた展開式でも確認できます。

全変動 $S_T =\|\mathbf{t}\|^2  = \sum_{i,j} (x_{ij} - \bar{\bar{x}})^2$ の中身を分解して展開します。

$$
\begin{aligned}
\sum_{i=1}^k \sum_{j=1}^{n_i} (x_{ij} - \bar{\bar{x}})^2
&= \sum_{i=1}^k \sum_{j=1}^{n_i}
\left\{
(\underbrace{\bar{x}_i - \bar{\bar{x}}}_{\text{シグナル}})
+
(\underbrace{x_{ij} - \bar{x}_i}_{\text{ノイズ}})
\right\}^2 \\
&= \underline{
\sum_{i=1}^k \sum_{j=1}^{n_i} (\bar{x}_i - \bar{\bar{x}})^2
}_{S_A}
+ \underline{
\sum_{i=1}^k \sum_{j=1}^{n_i} (x_{ij} - \bar{x}_i)^2
}_{S_E}
+ 2 \underline{
\sum_{i=1}^k \sum_{j=1}^{n_i}
(\bar{x}_i - \bar{\bar{x}})(x_{ij} - \bar{x}_i)
}_{\text{交差項}}
\end{aligned}
$$

ここで、最後の「交差項」が $0$ になれば $S_T = S_A + S_E$ が成り立ちます。
交差項の $\sum$ の順序を整理すると：

$$
\text{交差項} = 2 \sum_{i=1}^k \left\{ (\bar{x}_i - \bar{\bar{x}}) \underline{ \sum_{j=1}^{n_i} \left(x_{ij} - \bar{x}_i \right) }_{0} \right\} = 0
$$

内側の $\sum_{j=1}^{n_i} (x_{ij} - \bar{x}_i)$ は、「各グループ内での平均からの偏差の和」なので、平均の定義により必ず 0 になります。
これが、ベクトル表現における「直交性 ($\mathbf{a} \cdot \mathbf{e} = 0$)」の正体です。

#### **自由度と次元**

「自由度」とは、それぞれのベクトルが動くことのできる空間の次元のことです。
ここで扱うベクトル（$\mathbf{t}, \mathbf{a}, \mathbf{e}$）はすべて「成分の総和が0」であるため、$\mathbf{1} = (1, \dots, 1)^{\mathrm{T}} \in \mathbb{R}^n$ というベクトルに直交する超平面上に存在するという制約を共通して持っています。

- 全変動の自由度:全データ数 $n$ の空間。$\mathbf{1} \in \mathbb{R}^n$ ベクトルへの直交制約により次元は $n-1$ 。
$$\dim(\mathbf{t}) = n-1$$
- シグナルの自由度: グループ数 $k$ の空間。同様に $\mathbf{1} \in \mathbb{R}^k$ ベクトルへの直交制約により次元は $k-1$。
$$\dim(\mathbf{a}) = k-1$$
- ノイズの自由度: ノイズベクトル $\mathbf{e}$ は、全変動空間（次元 $n-1$）の中で、シグナル空間（次元 $k-1$）と直交する残りの空間（直交補空間）に存在。よってその次元は引き算で求まる。
$$\dim(\mathbf{e}) = (n-1) - (k-1) = n-k$$


#### **具体例で確認してみよう！**

##### **1. ベクトル演算と総和記号の計算比較**

Pythonを使って「ベクトル演算によるアプローチ」と「総和計算によるアプローチ」の両方を実装し、結果が完全に一致することを確かめましょう。

```python
import pandas as pd
import numpy as np

# 1. データの作成
data = {
    'Class_A': [82, 85, 88, 90, 85],
    'Class_B': [70, 75, 72, 78, 75],
    'Class_C': [60, 65, 62, 68, 65]
}
# データフレームの整形 (Tidy Data)
df_wide = pd.DataFrame(data)
df = df_wide.melt(var_name='Group', value_name='Score')

# 各種平均値の計算
grand_mean = df['Score'].mean()  # 全体平均
# transformを使って、各データに対応するグループ平均を並べたベクトルを作成
group_means = df.groupby('Group')['Score'].transform('mean')

display(df)
```

```python
# ========================================================
# アプローチ 1: ベクトル演算
# ========================================================
# NumPy配列としてベクトルを取得 (型をfloatに統一して計算精度を確保)
x_vec = df['Score'].values.astype(float)      # データベクトル x
group_mean_vec = group_means.values.astype(float) # グループ平均ベクトル

# 全体平均ベクトル (スカラーを引き算することでブロードキャストを利用)
# t_vec = x_vec - grand_mean でも同じですが、明示的にベクトルとして扱います
grand_mean_vec = np.full_like(x_vec, grand_mean)

# ベクトルの分解
t_vec = x_vec - grand_mean_vec      # 全変動ベクトル t
a_vec = group_mean_vec - grand_mean_vec # シグナルベクトル a
e_vec = x_vec - group_mean_vec      # ノイズベクトル e

# ノルムの2乗（内積）による平方和の計算
S_T_vec = np.sum(t_vec**2)
S_A_vec = np.sum(a_vec**2)
S_E_vec = np.sum(e_vec**2)

print("--- ベクトル演算による結果 ---")
print(f"全変動 (S_T): {S_T_vec:.2f}")
print(f"級間変動 (S_A): {S_A_vec:.2f}")
print(f"級内変動 (S_E): {S_E_vec:.2f}")
print(f"直交性の確認 (a・e): {np.dot(a_vec, e_vec):.3f}")
```

```python
# ========================================================
# アプローチ 2: 総和計算
# ========================================================
# 事前にグループごとの情報を辞書などで整理しておきます
# これにより、kやn_iといった変数の対応関係が明確になります
group_stats = df.groupby('Group')['Score'].agg(['count', 'mean'])
k = len(group_stats) # グループ数

print(f"総和記号の範囲確認:")
print(f"- グループ数 k = {k}")
print("- 各グループのデータ数 n_i:")
print(group_stats['count'])

# S_T: ΣΣ (x_ij - grand_mean)^2
# 全データと全体平均との差の2乗和
S_T_sum = np.sum((df['Score'] - grand_mean)**2)

# S_A: Σ n_i * (mean_i - grand_mean)^2
S_A_sum = 0
for group_name in group_stats.index:
    n_i = group_stats.loc[group_name, 'count']   # そのグループのデータ数 n_i
    mean_i = group_stats.loc[group_name, 'mean'] # そのグループの平均 mean_i
    S_A_sum += n_i * (mean_i - grand_mean)**2

# S_E: ΣΣ (x_ij - mean_i)^2
S_E_sum = 0
for group_name in group_stats.index:
    scores_i = df[df['Group'] == group_name]['Score'] # そのグループのデータ群 x_ij
    mean_i = group_stats.loc[group_name, 'mean']      # そのグループの平均 mean_i
    # グループ内の偏差平方和を計算して足し合わせる
    S_E_sum += np.sum((scores_i - mean_i)**2)

print("--- 総和計算による結果 ---")
print(f"全変動 (S_T): {S_T_sum:.2f}")
print(f"級間変動 (S_A): {S_A_sum:.2f}")
print(f"級内変動 (S_E): {S_E_sum:.2f}")
print(f"検算 (S_A + S_E): {S_A_sum + S_E_sum:.2f}")
```

```python
# ========================================================
# F値の算出と分散分析表の作成
# ========================================================

# 自由度とF値の計算
n = len(df)
dof_a = k - 1
dof_e = n - k
dof_t = n - 1

V_A = S_A_vec / dof_a
V_E = S_E_vec / dof_e
F_value = V_A / V_E

# 分散分析表（ANOVA Table）の作成と表示
anova_table = pd.DataFrame({
    '変動要因': ['級間 (Signal)', '級内 (Noise)', '全体 (Total)'],
    '平方和 (S)': [S_A_vec, S_E_vec, S_T_vec],
    '自由度 (df)': [dof_a, dof_e, dof_t],
    '平均平方 (V)': [V_A, V_E, np.nan],  # 全体の平均平方は通常計算しないためNaN
    'F値': [F_value, np.nan, np.nan]
})

# 見やすさのために小数点以下を丸める設定
pd.options.display.float_format = '{:.2f}'.format

print("--- 分散分析表 (ANOVA Table) ---")
display(anova_table.set_index('変動要因'))
```

##### **2. シグナルベクトルの可視化による自由に動ける次元としての自由度の理解**
グループ数が$k=3$の場合に、シグナルベクトル $\mathbf{a}$ を$k=3$次元空間に図示します。また、同様なデータが複数得られた場合のシグナルベクトルも同時に図示することで、シグナルベクトルがどのような次元の空間に存在するのか確認してみましょう。

```python
# ========================================================
#  幾何学的な視点：シグナル空間における帰無仮説 (H0) の分布と実データの比較
# ========================================================
import numpy as np
import plotly.graph_objects as go

# 1. 観測された実データのシグナル成分 (各グループ平均 - 全体平均)
effects = df.groupby('Group')['Score'].mean() - grand_mean
a1, a2, a3 = effects.loc['Class_A'], effects.loc['Class_B'], effects.loc['Class_C']
a_vec = np.array([a1, a2, a3])

# 2. 帰無仮説 (H0) 下での「あり得たかもしれない結果」の生成
# 「本当はグループ間に差がない」と仮定した場合の、ノイズによる平均値の変動をシミュレーションします。
np.random.seed(42)
num_sim = 300

# 級内分散 (V_E) から平均値の標準誤差 (SE) を算出
# ※ 演習Aで計算した V_E を使用。 1グループあたりのデータ数 n_i = 5
n_i = 5
se = np.sqrt(V_E / n_i)

# 中心を「原点 (0, 0, 0)」にして、ノイズの大きさ(se)で散らします。
# これが「F分布が想定しているノイズの霧」の正体です。
s_h0_1 = np.random.normal(0, se, num_sim)
s_h0_2 = np.random.normal(0, se, num_sim)
s_h0_3 = -s_h0_1 - s_h0_2 # x + y + z = 0 の制約

# 3. 制約平面 (x + y + z = 0) の作成
range_lim = max(abs(a1), abs(a2), abs(a3)) + 5
x_grid = np.linspace(-range_lim, range_lim, 20)
y_grid = np.linspace(-range_lim, range_lim, 20)
X, Y = np.meshgrid(x_grid, y_grid)
Z = -X - Y

fig = go.Figure()

# --- 各要素のプロット ---

# A. シグナル平面 (Signal Space)
fig.add_trace(go.Surface(
    x=x_grid, y=y_grid, z=Z,
    colorscale='Blues', opacity=0.2, showscale=False,
    name='Signal Space (df=k-1)', hoverinfo='skip'
))

# B. 帰無仮説 (H0) の霧：差がない場合にノイズが生み出す分布
# p値を計算する際の「物差し」となる分布です。
fig.add_trace(go.Scatter3d(
    x=s_h0_1, y=s_h0_2, z=s_h0_3,
    mode='markers',
    marker=dict(size=2, color='royalblue', opacity=0.3),
    name='H0 Distribution (Noise Only)',
    showlegend=True
))

# C. 観測された実データのシグナルベクトル
fig.add_trace(go.Scatter3d(
    x=[0, a_vec[0]], y=[0, a_vec[1]], z=[0, a_vec[2]],
    mode='lines', line=dict(color='red', width=8),
    name='Observed Signal (Real Data)',
    showlegend=True
))

# ベクトルの先端を指し示すコーン
fig.add_trace(go.Cone(
    x=[a_vec[0]], y=[a_vec[1]], z=[a_vec[2]],
    u=[a_vec[0]], v=[a_vec[1]], w=[a_vec[2]],
    sizemode="absolute", sizeref=1.5, anchor="tip",
    colorscale=[[0, 'red'], [1, 'red']], showscale=False, hoverinfo='skip'
))

# 観測データ点
fig.add_trace(go.Scatter3d(
    x=[a_vec[0]], y=[a_vec[1]], z=[a_vec[2]],
    mode='markers',
    marker=dict(size=8, color='red', symbol='circle'),
    name='Observed Data Point',
    showlegend=False
))

# D. 全体平均の軸
l_range = np.array([-range_lim, range_lim])
fig.add_trace(go.Scatter3d(
    x=l_range, y=l_range, z=l_range,
    mode='lines', line=dict(color='gray', width=2, dash='dash'),
    name='Grand Mean Axis', showlegend=True
))

# E. 原点
fig.add_trace(go.Scatter3d(
    x=[0], y=[0], z=[0],
    mode='markers', marker=dict(size=6, color='black'),
    name='Origin (Perfect H0)', showlegend=True
))

# レイアウト設定
fig.update_layout(
    title='Statistical Significance: Observed Signal vs H0 Distribution',
    scene=dict(
        xaxis_title='Group A Effect', yaxis_title='Group B Effect', zaxis_title='Group C Effect',
        aspectmode='cube'
    ),
    margin=dict(l=0, r=0, b=0, t=40),
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01)
)

fig.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# グラフのスタイル設定
sns.set(style="whitegrid")

# 1. データの作成（演習 A と同一のデータ）
data = {
    'Class_A': [82, 85, 88, 90, 85],
    'Class_B': [70, 75, 72, 78, 75],
    'Class_C': [60, 65, 62, 68, 65]
}
df_wide = pd.DataFrame(data)
df = df_wide.melt(var_name='Group', value_name='Score')

# 2. ライブラリ (statsmodels) による分散分析
# OLS (最小二乗法) モデルを作成し、分散分析表を出力します
model = ols('Score ~ Group', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("--- statsmodels による分散分析表 ---")
display(anova_table)

# 3. 検定結果の判定
p_value = anova_table.loc['Group', 'PR(>F)']
print(f"\nP値: {p_value:.10e}")

# 4. 視覚化：箱ひげ図
# データの分布とグループ間の差異を視覚的に確認します
plt.figure(figsize=(8, 5))
# FutureWarningへの対応: hueを指定し、不要な凡例を非表示に設定
sns.boxplot(x='Group', y='Score', data=df, palette="Set2", hue='Group', legend=False)
sns.stripplot(x='Group', y='Score', data=df, color='black', alpha=0.5)
plt.title('Analysis of Variance: Boxplot (Signal vs Noise)')
plt.show()
```
