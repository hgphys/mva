## **要点まとめ**
**Part I：座標と情報**

多変量解析の基礎として、データをベクトルとして捉え、そのベクトル同士の「関係性（絡まり合い）」を数値化する方法を学びます。

### **1. 質的データの関係：クロス集計**

質的データ（カテゴリ）同士の関係を俯瞰するには、それぞれのカテゴリーの組み合わせの出現頻度を表にまとめる **クロス集計（Cross Tabulation）** が有効です。

例えば、「クラス（A/B）」と「合否（合格/不合格）」の関係を見たい場合、元のデータは以下のような文字の入った表です。

| サンプル | クラス | 合否 |
| :---: | :---: | :---: |
| 1 | A | 合格 |
| 2 | B | 不合格 |
| ... | ... | ... |
<br>

これを集計すると、以下のようなクロス集計表（Aかつ合格は何人、といった頻度の表）になります。

| | 合格 | 不合格 | 計 |
|---|---|---|---|
| **クラスA** | 30 | 10 | 40 |
| **クラスB** | 20 | 20 | 40 |
| 計 | 50 | 30 | 80 |

<br>

### **2. 量的データの関係：共分散と相関行列**

量的データ（数値）同士の関係性は、**「分散共分散行列（Variance-Covariance Matrix）」** や **「相関行列（Correlation Matrix）」** として要約されます。

例えば、「広さ」と「家賃」のような量的データ（数値）の場合、単なる数値ではなく **「平均からのズレ（偏差）」** の列を作ってみると関係性が見えてきます。

| サンプル | 広さ $x$ | 家賃 $y$ | 偏差 $x - \bar{x}$ | 偏差 $y - \bar{y}$ | 偏差の積 |
| :---: | :---: | :---: | :---: | :---: | :---: |
| 1 | 50 | 10 | +10 (正) | +2 (正) | +20 |
| 2 | 30 | 6 | -10 (負) | -2 (負) | +20 |
| ... | ... | ... | ... | ... | ... |

<br>

このようにデータを作り変えると、量的データの関係性（相関）も、以下のような **「符号のクロス集計」** として捉えることができます。

- 正の相関: $(+, +)$ や $(- , -)$ の組み合わせが多い（積がプラスになる）。
- 負の相関: $(+, -)$ や $(- , +)$ の組み合わせが多い（積がマイナスになる）。

この「偏差の積の平均」こそが、以下で定義する共分散の正体です。

#### **(1) 分散共分散行列とは**

分散共分散行列とは、対角成分に「分散」、非対角成分に「共分散」を並べた行列です。データの広がりの大きさと方向を表します。

- **分散共分散行列$\Sigma$ の定義**<br>
サンプル数 $n$ の変数 $x_i$ と変数 $x_j$ の分散共分散行列の要素$(\Sigma)_{ij}=s_{ij}$（または $\sigma_{ij}$）は以下のように定義されます。
  - **対角成分 ($i = j$)：分散**<br>
  変数 $x_i$ 自身のばらつきを表します。共分散の式で $i=j$ としたものです。
$$s_{ii} = \frac{1}{n} \sum_{k=1}^{n} (x_{ki} - \bar{x}_i)^2$$

  - **非対角成分 ($i \neq j$)：共分散**<br>
  変数 $x_i$ と $x_j$ の関係性を表します。
$$s_{ij} = \frac{1}{n} \sum_{k=1}^{n} (x_{ki} - \bar{x}_i)(x_{kj} - \bar{x}_j)$$

- **行列とベクトルによる表記**<br>
中心化行列 $X_c$（各列から平均を引いた行列）を用いると、シンプルに記述できます。
$$\Sigma = \frac{1}{n} X_c^T X_c$$
この行列の対角成分は、各変数の偏差ベクトル（データベクトル $\mathbf{x}_i$ から平均ベクトル $\mathbf{\bar{x}}_i$ を引いたもの）のL2ノルム（長さ）の2乗に対応しています。
$$s_{ii} = \frac{1}{n} \|\mathbf{x}_i - \mathbf{\bar{x}}_i\|^2$$

<br>

#### **(2) 相関行列 $R$ とは**

共分散を各変数の標準偏差で割り、単位（スケール）の影響を排除した行列です。

- 要素（総和記号）による定義: <br>
変数 $x_i$ と変数 $x_j$ の相関係数 $r_{ij}$ は以下のようになります。
$$r_{ij} = \frac{s_{ij}}{\sqrt{s_{ii}}\sqrt{s_{jj}}} = \frac{\sum_{k=1}^{n} (x_{ki} - \bar{x}_i)(x_{kj} - \bar{x}_j)}{\sqrt{\sum_{k=1}^{n} (x_{ki} - \bar{x}_i)^2} \sqrt{\sum_{k=1}^{n} (x_{kj} - \bar{x}_j)^2}}$$

- 行列による定義: <br>
標準化行列 $Z$（平均0、分散1に変換した行列）を用いると、内積として記述できます。
$$R = \frac{1}{n} Z^T Z$$

<br>

#### **(3) 数理最適化・幾何学の視点：内積とコサイン**

統計的な「相関」は、幾何学（ベクトル）の言葉では **「内積（角度）」** として理解できます。

データを $n$ 次元のベクトル $\mathbf{x}, \mathbf{y}$ と見なし、これらを標準化（長さ1に正規化）して内積をとると、それは相関係数と完全に一致します。また、これは幾何学における **コサイン類似度（$\cos \theta$）** そのものです。

$$r_{xy} = \frac{\mathbf{x}\cdot\mathbf{y}}{||\mathbf{x}||~||\mathbf{y}||} = \cos \theta$$

つまり、 **「相関が高い」** とは **「ベクトルの向きが似ている（角度が近い）」** ということです。なお、分散共分散行列は、中心化データの内積であり、角度に加えて **「大きさ（分散）」** の情報も含むことになります。

<br>

## **Python 演習課題**

### **mva_03-A**
✅ **目的**: データの「広がり（分布）」を行列演算で捉え、それを考慮した公平な距離尺度である「マハラノビス距離」を理解します。共分散行列が作る「歪んだ座標空間」において、距離がどのように定義されるかを幾何学的（誤差楕円）に確認しましょう。

#### 🖊 **【数理理解】データの分布とマハラノビス距離**

データ空間では、変数のスケールや相関（情報の偏り）によって、距離が公平に測れません。そこで、データが持つ情報（分散・共分散）を使って座標を補正する、という考え方を導入します。

##### **1. 情報の形（分散共分散行列 $\Sigma$）**
データの分布は、一般に原点（平均）を中心に楕円状に広がります。この「楕円の形」を決めるのが、分散共分散行列 $\Sigma$ です。
データ行列 $X_c$（中心化済み）から、行列演算によって以下のように導かれます。

$$\Sigma = \frac{1}{n} X_c ^\top X_c\quad \quad \left(~X_c = X - \mathbf{1}_n \bar{\mathbf{x}}^{\mathrm{T}}~\right)$$
- **対角成分** $(\Sigma)_{ii} =s_{ii}$ : 分散。各軸方向の広がり。
- **非対角成分** $(\Sigma)_{ij} =s_{ij}$ : 共分散。軸同士の傾き。

<br>

##### **2. 情報の物差し（マハラノビス距離）**
この「楕円状の座標空間」において、中心からの距離を公平に測るための尺度がマハラノビス距離です。 通常のユークリッド距離が「物理的な距離」であるのに対し、マハラノビス距離はデータの分布を考慮した「統計的な距離（確率的な遠さ）」です。マハラノビス距離の2乗は以下のように定義されます。

$$D_M(\mathbf{x})^2 = (\mathbf{x} - \mathbf{\bar{x}})^\top \Sigma^{~-1} (\mathbf{x} -\mathbf{\bar{x}})$$

マハラノビス距離が $1$ となる領域は、中心から 標準偏差 1個分 ($1\sigma$) 離れた等高線を表します。

- **無相関な場合の距離（標準化との関係）**<br>
2変数が互いに無相関（共分散が0）の場合、分散共分散行列 $\Sigma$ は対角行列になります。
$$\Sigma = \begin{pmatrix} s_{11} & 0 \\ 0 & s_{22} \end{pmatrix} \quad \Rightarrow \quad \Sigma^{-1} = \begin{pmatrix} \frac{1}{s_{11}} & 0 \\ 0 & \frac{1}{s_{22}} \end{pmatrix}$$
このとき、逆行列 $\Sigma^{~-1}$ は分散の逆数を対角成分に持つ行列となり、マハラノビス距離の2乗は以下のように変形できます。
$$D_M(\mathbf{x})^2 = \frac{(x_1 - \bar{x}_1)^2}{s_{11}} + \frac{(x_2 - \bar{x}_2)^2}{s_{22}} = \left(\frac{x_1 - \bar{x}_1}{\sqrt{s_{11}}}\right)^2 + \left(\frac{x_2 - \bar{x}_2}{\sqrt{s_{22}}}\right)^2=z_1 ^2 +z_2 ^2$$
これは、各変数を標準偏差で割って標準化（Z値化）した後の、ユークリッド距離の2乗和に他なりません。
つまり、「逆行列 $\Sigma^{~-1}$ をかける」という操作は、本質的に「スケーリング行列 $S$ で割って標準化する」ことの一般化なのです。

- **行列による座標の補正（$\Sigma^{-1}$ の役割）**<br>
分散共分散行列の逆行列は、歪んだデータ空間を「正円」に戻す補正フィルタの役割を果たします。
平均を原点とした2次元データ $(x, y)$ において、**距離が 1 となる点の軌跡（方程式）** を比較すると、その役割が明確になります。<br>
| 距離の種類 | 前提 | 補正行列（重み） | 距離 $D^2=1$ の方程式（軌跡） |
| :--- | :--- | :--- | :--- |
| **ユークリッド距離** | 変数は対等・無関係 | 単位行列 $I$ | $x^2 + y^2 = 1$<br>（正円） |
| **標準ユークリッド距離** | スケールが違う | 分散の逆数 $S^{-1}$ | $\frac{x^2}{s_x^2} + \frac{y^2}{s_y^2} = 1$<br>（軸に平行な楕円） |
| **マハラノビス距離** | 相関がある | 逆行列 $\Sigma^{-1}$ | $a x^2 + 2b xy + c y^2 = 1$ <br>（回転した楕円） |

<br>

#### **具体例で確認してみよう！**
##### **1. 行列演算での分散共分散行列と相関行列の計算**

第2回で使用したスプレッドシートデータを読み込み、行列演算で分散共分散行列と相関行列を計算できることを確認します。

読み込み先URL:
[https://docs.google.com/spreadsheets/d/1v_FgNUblx_DpXIzmvf9pl1F_eCqMzu10XVjL7YgTfUQ](https://docs.google.com/spreadsheets/d/1v_FgNUblx_DpXIzmvf9pl1F_eCqMzu10XVjL7YgTfUQ)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import seaborn as sns

# ---------------------------------------------------------
# 1. データの読み込み
# ---------------------------------------------------------
# Googleスプレッドシートからデータを読み込む
sheet_url = "https://docs.google.com/spreadsheets/d/1v_FgNUblx_DpXIzmvf9pl1F_eCqMzu10XVjL7YgTfUQ/export?format=csv"
df = pd.read_csv(sheet_url, index_col=0)

# フォント問題を回避するため列名を英語に変更
df.columns = ['x1', 'x2']
X = df.to_numpy()
n = X.shape[0]

print("--- 読み込んだデータ ---")
display(df.head())

# ---------------------------------------------------------
# 2. 行列演算による統計量の算出
# ---------------------------------------------------------
# (1) 平均ベクトル
mean_vec = np.mean(X, axis=0)

# (2) 中心化 (Broadcasting を利用)
# データベクトル x_i から 平均ベクトル mean_vec を引く
Xc = X - mean_vec

# (3) 分散共分散行列 (Sigma = 1/n * Xc.T @ Xc)
# 偏差ベクトルの内積の平均
Sigma = (Xc.T @ Xc) / n

# (4) 標準化
# 中心化行列を標準偏差で割る
std_vec = np.std(X, axis=0, ddof=0)
Z = Xc / std_vec

# (5) 相関行列 (R = 1/n * Z.T @ Z)
# 標準化ベクトルの内積の平均
R = (Z.T @ Z) / n

print("\n--- 分散共分散行列 (Sigma) ---\n", Sigma)
print("\n--- 相関行列 (R) ---\n", R)
```

##### **2. マハラノビス距離を基準としたデータ分布の可視化**

元のデータ空間と標準化後の空間において、データの広がりがどのような「楕円」で表されるかを比較してみましょう。

```python
# ---------------------------------------------------------
# 3. 可視化：誤差楕円の描画関数
# ---------------------------------------------------------
def plot_cov_ellipse(cov, pos, n_std=1, ax=None, **kwargs):
    """
    共分散行列から誤差楕円を描画する関数
    """
    if ax is None:
        ax = plt.gca()

    # 固有値・固有ベクトルを計算（楕円の長軸・短軸と回転角を求める）
    eigvals, eigvecs = np.linalg.eigh(cov)

    # 固有値を降順にソート
    order = eigvals.argsort()[::-1]
    eigvals, eigvecs = eigvals[order], eigvecs[:, order]

    # 回転角を計算 (ラジアン -> 度)
    theta = np.degrees(np.arctan2(*eigvecs[:,0][::-1]))

    # 楕円の幅と高さを計算 (2 * n_std * 標準偏差)
    # 固有値は分散に相当するため、ルートをとって標準偏差にする
    width, height = 2 * n_std * np.sqrt(eigvals)

    # 楕円オブジェクトの作成
    ell = Ellipse(xy=pos, width=width, height=height, angle=theta, **kwargs)
    ax.add_patch(ell)
    return ell

# ---------------------------------------------------------
# 4. プロットによる比較
# ---------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# --- (左) 元データ空間：分散共分散行列による楕円 ---
axes[0].scatter(X[:, 0], X[:, 1], alpha=0.5, label='Data')
# 1シグマの楕円を描画 (n_std=1)
plot_cov_ellipse(Sigma, pos=mean_vec, n_std=1, ax=axes[0],
                 edgecolor='red', facecolor='none', linewidth=2, label='1-sigma Range (Sigma)')
# 重心をプロット
axes[0].scatter(mean_vec[0], mean_vec[1], color='red', marker='x', s=100, label='Mean')

axes[0].set_title("Original Space (Covariance Matrix)")
axes[0].set_xlabel("Area (x1)")
axes[0].set_ylabel("Rent (x2)")
axes[0].set_xlim([-100, 1000])
axes[0].set_ylim([-100, 1000])
axes[0].legend()

# --- (右) 標準化空間：相関行列による楕円 ---
axes[1].scatter(Z[:, 0], Z[:, 1], alpha=0.5, color='green', label='Standardized Data')
# 標準化後の中心は原点(0,0)、共分散行列は相関行列Rになる
# 1シグマの楕円を描画 (n_std=1)
plot_cov_ellipse(R, pos=(0, 0), n_std=1, ax=axes[1],
                 edgecolor='red', facecolor='none', linewidth=2, label='1-sigma Range (Correlation)')

# 参考：無相関の場合の円 (半径1の単位円)
circle = plt.Circle((0, 0), 1, color='blue', linestyle='--', fill=False, label='Uncorrelated (Unit Circle)')
axes[1].add_patch(circle)

axes[1].set_title("Standardized Space (Correlation Matrix)")
axes[1].set_xlabel("z-score (x1)")
axes[1].set_ylabel("z-score (x2)")
axes[1].legend()
axes[1].set_xlim(-5, 5)
axes[1].set_ylim(-5, 5)

plt.tight_layout()
plt.show()
```

### **mva_03-B**

✅ **目的**: ``pandas`` や ``seaborn`` ライブラリを使い、データフレームから一括で相関行列を算出し、ヒートマップとして可視化する実務的な手順を習得しましょう。

#### 💻 **【実装・応用】 ライブラリによる相関分析と可視化**

実際の分析では、数多くの変数間の関係性を瞬時に把握する必要があります。
``pandas`` の ``corr()`` メソッドは、データフレーム内のすべての数値列ペアについて相関係数を一括計算します。
また、``seaborn`` の ``heatmap`` を使うことで、数値の羅列である相関行列を「色の濃淡」として直感的に理解できる図に変換できます。

詳細な仕様については、以下の公式ドキュメントを参照してください。

- ``pandas`` 公式ドキュメント: [DataFrame.corr](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html)
- ``seaborn`` 公式ドキュメント: [seaborn.heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html)



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. データの読み込み
sheet_url = "https://docs.google.com/spreadsheets/d/1KhE_eZcLvkhlnnN5d5TYkAwUEp33qwjz7qP9OqG91x0/export?format=csv"
df = pd.read_csv(sheet_url, index_col=0)

print("--- 読み込んだデータ ---")
display(df.head())

# 2. 相関行列の計算
# 数値データのみを抽出して相関行列を計算します
# numeric_only=True を指定することで、非数値列（文字列など）を自動的に除外できます
corr_matrix = df.corr(numeric_only=True)

print("\n--- 相関行列 (Correlation Matrix) ---")
display(corr_matrix)

# 3. ヒートマップによる可視化
plt.figure(figsize=(8, 6))
# annot=True: 数値をセルに表示
# cmap='coolwarm': 正の相関を赤、負の相関を青で表現
# vmin=-1, vmax=1: カラースケールの範囲を相関係数の定義域に固定
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
