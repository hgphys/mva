## **要点まとめ**

**Part V：分離の最大化**

「判別分析」では、数値データを用いてグループを分類しました。今回学習する **数量化Ⅱ類（Quantification Theory Type II）** は、説明変数が **「質的データ（カテゴリ）」** である場合に、サンプルがどのグループ（目的変数）に属するかを予測・判別する手法です。

### **1. 数量化Ⅱ類の位置付け**

数理的には、**「質的データをダミー変数に変換し、線形判別分析を適用する手法」** と捉えることができます。

| 手法 | 説明変数 ($X$) | 目的変数 ($y$) | 目的・アプローチ |
| :--- | :--- | :--- | :--- |
| **重回帰分析** | 数値 | 数値 | 誤差の最小化（回帰） |
| **数量化Ⅰ類** | **カテゴリ** | 数値 | 誤差の最小化（回帰） |
| **判別分析** | 数値 | **カテゴリ** | 分離の最大化（分類） |
| **数量化Ⅱ類** | **カテゴリ** | **カテゴリ** | **分離の最大化（分類）** |

<br>

### **2. 判別得点の定義**

多次元の質的データを1次元の情報に凝縮するため、各アイテムのカテゴリに割り当てられた重み（カテゴリ・スコア） $w$ を掛けて足し合わせます。これを **判別得点 (Discriminant Score)** $z_i$ と呼びます。

$$z_i = \sum_{j=1}^{d} \sum_{k=1}^{c_j} w_{jk} x_{ijk} = \mathbf{w}^\top \mathbf{x}_i$$

- $d$: アイテム（項目）の総数
- $c_j$: アイテム $j$ におけるカテゴリの総数
- $\mathbf{x}_i$: サンプル $i$ をダミー変数化したベクトル
- $\mathbf{w}$: すべてのカテゴリに対するスコア（重み）のベクトル

この式は、判別分析と同様に、質的情報をダミー変数空間におけるベクトル $\mathbf{x}_i$ と重みベクトル $\mathbf{w}$ の 内積 によって「数量化」していることを意味します。

**【注意】ダミー変数トラップと制約条件**

数量化Ⅰ類と同様に、全てのカテゴリをダミー変数として用いると、各アイテム内のダミー変数の合計が常に $1$（定数）となり、変数間に完全な多重共線性（線形従属）が生じます。このため、後述する全分散行列 $S_T$ が正則（フルランク）ではなくなり、逆行列が計算できなくなります。これを避けるため、一般的には「各アイテムから1つのカテゴリを除外する」か、「アイテムごとにカテゴリ・スコアの平均を $0$ にする」といった制約条件を課して解を求めます。

<br>

### **3. 最適化のロジック：相関比の最大化と数理的解法**

数量化Ⅱ類では、射影後の世界（得点 $z$ の分布）において、グループを最も鮮明に分ける方向 $\mathbf{w}$ を探索します。

#### **(1) グループ間の離れ具合（群間分散の最大化）**

各グループの平均得点を $\bar{z}_k$ としたとき、グループ同士の離れ具合を 群間分散 $\sigma_B^2$ で評価します。これを最大化することで、集団同士を遠ざけます。

$$\sigma_B^2 = \sum_{k=1}^g \frac{n_k}{n} (\bar{z}_k - \bar{z})^2 = \mathbf{w}^\top S_B \mathbf{w}$$

ここで、$g$はグループ数（クラス数）、$n_k$はグループ $k$ に属するサンプル数。
$S_B$ はダミー変数空間における **群間分散行列** であり、次のように定義されます（$\bar{\mathbf{x}}_k$ はグループ平均ベクトル、$\bar{\mathbf{x}}$ は全体平均ベクトル）。
$$S_B = \sum_{k=1}^g \frac{n_k}{n} (\bar{\mathbf{x}}_k - \bar{\mathbf{x}})(\bar{\mathbf{x}}_k - \bar{\mathbf{x}})^\top$$

<br>

#### **(2) 全体のばらつき（全分散の固定）**

得点 $z$ の全体のばらつきを 全分散 $\sigma_T^2$ で評価します。判別効率を測る際、全体のばらつきに対してグループ間の違いが占める割合を評価するため、全分散を基準とします。

$$\sigma_T^2 = \frac{1}{n} \sum_{i=1}^n (z_i - \bar{z})^2 = \mathbf{w}^\top S_T \mathbf{w}$$

ここで、$S_T$ はダミー変数空間における **全分散行列（総散布行列）** であり、次のように定義されます。

$$S_T = \frac{1}{n} \sum_{i=1}^n (\mathbf{x}_i - \bar{\mathbf{x}})(\mathbf{x}_i - \bar{\mathbf{x}})^\top$$

<br>

#### **(3) 最適化の指標：相関比 $\eta^2$ と一般化固有値問題**

全体の変動のうち、どれだけグループの違いで説明できるかを表す指標を **相関比（Correlation Ratio）** $\eta^2$ と呼び、これを最大化することがゴールです。

$$J(\mathbf{w}) = \frac{\sigma_B^2}{\sigma_T^2} = \frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_T \mathbf{w}} $$

この $J(\mathbf{w}) $ を最大化する $\mathbf{w}$ は、以下の 一般化固有値問題 を解くことに帰着します。

$$S_T^{-1} S_B \mathbf{w} = \lambda \mathbf{w}$$

この最大固有値 $\lambda$（＝最大相関比 $\eta^2$）に対応する固有ベクトル $\mathbf{w}$ が、データを最も鮮明に判別できる軸となります。

<br>


## **Python 演習課題**

### **mva_11-A**

✅  **目的**: 具体的な行列 $S_T$ と $S_B$ を用いて固有値問題を解き、数理的に導かれた「最適な数量化ベクトル（カテゴリ・スコア）」が、カテゴリデータの判別にどのように寄与するかを体験します。

#### 🖊 **数理理解：固有値問題の導出ステップ**

質的データをダミー変数化した2次元の空間において、以下の数値例を用いて最適なスコアベクトル $\mathbf{w}$ を導出します。

**設定値**：
- クラス1（継続）とクラス2（退会）のダミー変数の平均値（反応率）の差が以下とします。
$$\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2 = \begin{pmatrix} 0.4 \\ 0.4 \end{pmatrix}$$
このとき、簡略化された群間分散行列 $S_B$ は次のように計算されます。

$$S_B = (\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)(\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)^\top = \begin{pmatrix} 0.4 \\ 0.4 \end{pmatrix} \begin{pmatrix} 0.4 & 0.4 \end{pmatrix} = \begin{pmatrix} 0.16 & 0.16 \\ 0.16 & 0.16 \end{pmatrix}$$

- 全分散行列を $S_T = \begin{pmatrix} 0.2 & 0.1 \\ 0.1 & 0.2 \end{pmatrix}$ とします。


<br>

**Step 1: 逆行列 $S_T^{-1}$ の計算**

行列式 $|S_T| = (0.2 \times 0.2) - (0.1 \times 0.1) = 0.03$ より、

$$S_T^{-1} = \frac{1}{0.03} \begin{pmatrix} 0.2 & -0.1 \\ -0.1 & 0.2 \end{pmatrix}$$

<br>

**Step 2: 行列 $A = S_T^{-1} S_B$ の合成**

$$A = \frac{1}{0.03} \begin{pmatrix} 0.2 & -0.1 \\ -0.1 & 0.2 \end{pmatrix} \begin{pmatrix} 0.16 & 0.16 \\ 0.16 & 0.16 \end{pmatrix} = \begin{pmatrix} 16/30 & 16/30 \\ 16/30 & 16/30 \end{pmatrix}$$

<br>

**Step 3: 固有値 $\lambda$ と固有ベクトル $\mathbf{w}$ の導出**

固有方程式 $|A - \lambda I| = 0$ を解くと：

$$\begin{vmatrix} 16/30 - \lambda & 16/30 \\ 16/30 & 16/30 - \lambda \end{vmatrix} = 0 \implies (16/30 - \lambda)^2 - (16/30)^2 = 0 \implies \lambda \left( \lambda - \frac{32}{30} \right) = 0$$

最大固有値 $\lambda = 32/30$ に対する固有ベクトル $\mathbf{w}$ は、$(A - \lambda I)\mathbf{w} = \mathbf{0}$ より：

$$\begin{pmatrix} -16/30 & 16/30 \\ 16/30 & -16/30 \end{pmatrix} \begin{pmatrix} w_1 \\ w_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix} \implies w_1 = w_2$$

したがって、最適なカテゴリ・スコアの比は $\mathbf{w}=[1, 1]^\top$ となります。

<br>

#### **具体例で確認してみよう！**

以下のコードを実行して、ダミー変数化されたデータに対し、固有値問題を解くことでカテゴリごとの「判別スコア」が算出される流れを確認しましょう。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from ipywidgets import interactive_output, FloatSlider, VBox, HTML
from IPython.display import display
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

# ---------------------------------------------------------
# ### mva_11-A：インタラクティブ演習（手計算の検証）
# ---------------------------------------------------------

# 1. カテゴリデータ（0/1）の生成
# [Δx=(0.4, 0.4)] に近い設定
def generate_quantification_data():
    np.random.seed(42)
    n_samples = 200 # 各グループのサンプル数

    # 理論値通りの平均(0.3 vs 0.7)を持つデータを生成し、
    # シミュレーション結果が理論解(45度)と一致するように調整します

    # Group 0: Mean 0.3 (30%が1, 70%が0)
    n_ones_0 = int(n_samples * 0.3)
    c1_0 = np.array([1]*n_ones_0 + [0]*(n_samples-n_ones_0)); np.random.shuffle(c1_0)
    c2_0 = np.array([1]*n_ones_0 + [0]*(n_samples-n_ones_0)); np.random.shuffle(c2_0)
    X0 = np.column_stack([c1_0, c2_0])

    # Group 1: Mean 0.7 (70%が1, 30%が0)
    n_ones_1 = int(n_samples * 0.7)
    c1_1 = np.array([1]*n_ones_1 + [0]*(n_samples-n_ones_1)); np.random.shuffle(c1_1)
    c2_1 = np.array([1]*n_ones_1 + [0]*(n_samples-n_ones_1)); np.random.shuffle(c2_1)
    X1 = np.column_stack([c1_1, c2_1])

    X = np.vstack([X0, X1]).astype(float)
    y = np.hstack([np.zeros(n_samples), np.ones(n_samples)])
    return X, y

X_raw, y = generate_quantification_data()

# 2. 理論解の定義 (w1=w2 のとき θ=45度)
theoretical_theta = 45.0

# 3. 描画関数
def update_plot(theta_deg):
    theta = np.radians(theta_deg)
    # 重みベクトル w = [cosθ, sinθ]
    v = np.array([np.cos(theta), np.sin(theta)])
    # 判別得点 z = w1*x1 + w2*x2
    # x1, x2 は 0か1 なので、z は 0, w1, w2, w1+w2 のいずれか近辺の離散値になる
    projected_z = np.dot(X_raw, v)

    # 相関比 (η^2) の計算
    z_mean_all = np.mean(projected_z)
    z_mean0 = np.mean(projected_z[y == 0])
    z_mean1 = np.mean(projected_z[y == 1])
    n = len(y)
    n0 = np.sum(y == 0)
    n1 = np.sum(y == 1)

    ss_b = (n0 * (z_mean0 - z_mean_all)**2 + n1 * (z_mean1 - z_mean_all)**2) / n
    ss_t = np.var(projected_z)
    eta_sq = ss_b / ss_t if ss_t > 0 else 0

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
    colors = ['#14b8a6', '#f59e0b'] # Teal and Orange

    # --- 左側: ダミー変数空間 (2D) ---
    # 実際は (0,0), (1,0), (0,1), (1,1) の4点に全サンプルが集中している
    # 密度（サンプル数）を視覚化するため、描画時のみ微小なノイズ（Jitter）を加える
    jitter = np.random.normal(0, 0.04, X_raw.shape)
    X_jitter = X_raw + jitter

    for label, color in zip([0, 1], colors):
        ax1.scatter(X_jitter[y == label, 0], X_jitter[y == label, 1],
                    alpha=0.6, c=color, label=f'Group {label}', edgecolors='w', s=40)

    # 重みベクトル w の軸
    line_len = 1.2
    ax1.plot([-v[0]*0.1, v[0]*line_len], [-v[1]*0.1, v[1]*line_len],
             'k--', linewidth=1.5, label=f'Weight Vector w ({theta_deg:.1f}°)')

    # 理論解 (45度) のガイド
    v_opt = np.array([np.cos(np.radians(theoretical_theta)), np.sin(np.radians(theoretical_theta))])
    ax1.plot([-v_opt[0]*0.1, v_opt[0]*line_len], [-v_opt[1]*0.1, v_opt[1]*line_len],
             'red', linewidth=6, alpha=0.15, label='Optimal w (45°)')

    ax1.set_xlim(-0.2, 1.2); ax1.set_ylim(-0.2, 1.2); ax1.set_aspect('equal')
    ax1.set_title("Input: Categorical Data (with Jitter)")
    ax1.set_xlabel("Category x1 (0 or 1)")
    ax1.set_ylabel("Category x2 (0 or 1)")
    ax1.set_xticks([0, 1]); ax1.set_yticks([0, 1])
    ax1.legend(loc='upper left'); ax1.grid(True, alpha=0.2)

    # --- 右側: 数量化された得点 z の分布 ---
    # 判別得点 z は離散値をとるが、分布の重なり（分離度）を見るために KDE を表示
    for label, color in zip([0, 1], colors):
        # KDE（カーネル密度推定）は離散データの「密度の傾向」を示す補助線
        sns.kdeplot(projected_z[y == label], ax=ax2, color=color, fill=True, alpha=0.2,
                    bw_adjust=0.5, label=f'Group {label}')

        # 実際のサンプル位置（4通りの離散値）をラグプロットで表示
        # 0/1の組み合わせによる得点の位置が、垂直線として現れる
        ax2.plot(projected_z[y == label], np.full_like(projected_z[y == label], -0.05 - label*0.05),
                 '|', color=color, alpha=0.6, markersize=20)

    ax2.set_title(f"Score Distribution (η² = {eta_sq:.4f})")
    ax2.set_xlabel("Discriminant Score z (Sum of Weights)")
    ax2.set_ylabel("Density / Frequency")
    ax2.set_xlim(-0.5, 1.8)
    ax2.legend()

    plt.tight_layout()
    plt.show()

# 4. インタラクティブUIの構築
angle_slider = FloatSlider(
    min=0, max=180, step=1.0, value=0,
    description='Angle [deg]',
    continuous_update=True,
    layout={'width': '500px'}
)

interactive_ui = interactive_output(update_plot, {'theta_deg': angle_slider})

display(HTML("""
<div style="background-color: #f0f7ff; padding: 15px; border-radius: 10px; margin-bottom: 10px;">
    スライダーを動かして 45° 付近でヒストグラムが綺麗に分かれることを確認してください。
</div>
"""))
display(VBox([angle_slider, interactive_ui]))
```

### **mva_11-B**

✅  **目的**: Pythonのライブラリ（``scikit-learn``）を用いて線形判別分析（LDA）を実装し、ダミー変数化された質的データ に対する決定境界がどのように引かれるかを視覚的に理解します。

#### 💻 **【実装・応用】ライブラリによる数量化Ⅱ類の実装**

実務において数量化Ⅱ類を行う際は、質的データをダミー変数に変換した後、``scikit-learn`` ライブラリの ``LinearDiscriminantAnalysis`` クラスを適用するのが一般的です。
今回は、具体的な会員データのサンプルを用い、ダミー変数トラップへの対応（多重共線性の回避）を行った上で、決定境界を描画し、ロジスティック回帰と比較を行います。

詳細な仕様については、以下の公式ドキュメントを参照してください。

- ``sklearn`` 公式ドキュメント: ``LinearDiscriminantAnalysis``

#### **具体例で確認してみよう！**

あるWebサービスの「有料会員登録 (Target)」を、「性別 (Gender)」と「利用頻度 (Usage)」という2つの質的変数から予測するシナリオです。

データの前処理: カテゴリ変数をOne-Hot Encodingし、冗長な変数を削除（drop_first=True）します。

モデル学習: LDAとロジスティック回帰の両方で学習します。

比較と可視化: 決定境界を描画し、両手法が「どのカテゴリの組み合わせを有料会員と判定するか」の違いを確認します。


読み込み先URL: [https://docs.google.com/spreadsheets/d/1Bdw4D1wz9q2KeAqUXNC3856r6Ux7SsIs7m8j1-acqyg](https://docs.google.com/spreadsheets/d/1Bdw4D1wz9q2KeAqUXNC3856r6Ux7SsIs7m8j1-acqyg/edit?usp=sharing)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

# ---------------------------------------------------------
# 1. データの読み込み (Google スプレッドシート)
# ---------------------------------------------------------
# 共有リンクの末尾を "/export?format=csv" に変更して読み込みます
# 例: https://docs.google.com/spreadsheets/d/.../export?format=csv
sheet_url = "https://docs.google.com/spreadsheets/d/1Bdw4D1wz9q2KeAqUXNC3856r6Ux7SsIs7m8j1-acqyg/export?format=csv"

try:
    df = pd.read_csv(sheet_url)
    print("Googleスプレッドシートからデータを読み込みました。")
except Exception:
    print("スプレッドシートの読み込みに失敗、またはURLが未設定です。")

print("\n--- 1. 元データ (最初の5行) ---")
display(df.head())

# ---------------------------------------------------------
# 2. One-Hot Encoding と ダミー変数トラップへの対応
# ---------------------------------------------------------
# drop_first=True にすることで、各カテゴリの最初の項目(Gender_F, Usage_Highなど)を除外し、
# 多重共線性（ダミー変数トラップ）を回避します。
# 残った変数が「1」ならそのカテゴリ、「0」なら基準カテゴリ（除外された方）を意味します。
X = pd.get_dummies(df[['Gender', 'Usage']], drop_first=True, dtype=int)
y = df['Target']

print("\n--- 2. ダミー変数化 (drop_first=True) ---")
display(X.head())
print(f"変数の意味: {X.columns[0]} (1=Male, 0=Female), {X.columns[1]} (1=Low, 0=High)")

# ---------------------------------------------------------
# 3. モデルの学習と係数の比較
# ---------------------------------------------------------
# 数量化II類 (LDA)
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)

# ロジスティック回帰 (比較用)
lr = LogisticRegression()
lr.fit(X, y)

# 係数（スコア）の比較表
coef_df = pd.DataFrame({
    'LDA Coef': lda.coef_[0],
    'Logistic Coef': lr.coef_[0]
}, index=X.columns)

print("\n--- 3. 判別スコア係数の比較 ---")
display(coef_df)
print("※ LDA係数の絶対値が大きいほど、そのカテゴリが判別に強く寄与しています。")

# ---------------------------------------------------------
# 4. 決定境界の可視化
# ---------------------------------------------------------
# 2次元のダミー変数空間 (Gender_M, Usage_Low) 上での決定境界を描画
# 座標は (0,0), (1,0), (0,1), (1,1) の4点のみをとります

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# グリッドの作成
xx, yy = np.meshgrid(np.arange(-0.5, 1.5, 0.02), np.arange(-0.5, 1.5, 0.02))

# データ点に少しノイズを加えて重なりを防ぐ (Jitter)
X_jitter = X.copy().astype(float)
X_jitter += np.random.normal(0, 0.05, X_jitter.shape)

models = [lda, lr]
titles = ["Quantification II (LDA)", "Logistic Regression"]

for ax, model, title in zip([ax1, ax2], models, titles):
    # 警告回避のため、予測用データをDataFrameに変換（学習時と同じ列名を使用）
    mesh_df = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=X.columns)

    Z = model.predict(mesh_df)
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.2, cmap=plt.cm.coolwarm)

    # 決定境界線
    ax.contour(xx, yy, Z, levels=[0.5], colors='black', linewidths=2)

    # データ点のプロット
    # Target=0 (Blue), Target=1 (Red)
    scatter = ax.scatter(X_jitter.iloc[:, 0], X_jitter.iloc[:, 1], c=y,
                         cmap=plt.cm.coolwarm, edgecolors='k', s=50, alpha=0.6)

    ax.set_title(title)
    ax.set_xlabel(X.columns[0])
    ax.set_ylabel(X.columns[1])
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
