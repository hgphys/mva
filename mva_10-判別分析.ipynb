## **要点まとめ**

**Part V：分離の最大化**

ロジスティック回帰分析では「確率（尤度）の最大化」を目指しましたが、今回の **線形判別分析（Linear Discriminant Analysis: LDA）** は、データをある方向に投影した際の「グループ間の分離度」を最大化するという幾何学的・統計的なアプローチをとります。
### **1. 判別分析とは**

判別分析の目的は、説明変数から目的変数（クラス）を予測するための最適な境界線（判別関数）を構築することです。データを行列形式で整理すると、以下のようになります。

| | 目的変数 (クラス)| 説明変数1 | ... | 説明変数d |
|---|:---:|:---:|:---:|:---:|
| **サンプル1** | $y_{1}$ | $x_{11}$  | ... | $x_{1d}$ |
| **サンプル2** | $y_{2}$ | $x_{21}$ | ... | $x_{2d}$ |
| ... | ... | ... | ... | ... |
| **サンプルn** | $y_{n}$ | $x_{n1}$ | ... | $x_{nd}$ |

- **目的変数**: $y_i \in \{C_1, C_2\}$ （ここでは2クラスの場合を想定）
- **説明変数ベクトル**: $\mathbf{x}_i = [x_{i1}, x_{i2}, \dots, x_{id}]^\top \in \mathbb{R}^d$
- **各クラスの集合**: サンプル $i$ のうち $y_i = C_k$ であるインデックスの集合を $I_k$、その要素数（サンプル数）を $n_k$ とします。

<br>

### **2. 判別得点の定義**

多次元のデータ $\mathbf{x}_i$ を1次元の情報に凝縮するため、各変数に重み $w$ を掛けて足し合わせます。これを **判別得点 (Discriminant Score)** $z_i$ と呼びます。

$$z_i = w_1 x_{i1} + w_2 x_{i2} + \dots + w_d x_{id} = \mathbf{w}^\top \mathbf{x}_i$$

この式は、幾何学的にはデータ $\mathbf{x}_i$ と重みベクトル $\mathbf{w}$ の **内積** を計算していることに他なりません。

<br>

### **3. フィッシャーの判別基準：最適化のロジックと数理的解法**

フィッシャーは、射影後の世界（得点 $z$ の分布）において、以下の2つの条件を同時に満たす方向 $\mathbf{w}$ が最適であると考えました。

#### **(1) クラス平均の離れ具合（グループ間分散の最大化）**

各クラスの中心を $\bar{\mathbf{x}}_k$ としたとき、射影後の中心 $\bar{z}_k = \mathbf{w}^\top \bar{\mathbf{x}}_k$ の離れ具合を グループ間分散 $\sigma_B^2$ で評価します。これを最大化することで、集団同士を遠ざけます。

$$\sigma_B^2 = (\bar{z}_1 - \bar{z}_2)^2 = \mathbf{w}^\top S_B \mathbf{w}$$

ここで、$S_B$ は射影前の空間における **群間分散行列** であり、次のように定義されます。

$$S_B = (\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)(\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)^\top$$

<br>

#### **(2) クラス内の凝集度（グループ内分散の最小化）**

各クラス内部での得点の散らばりを グループ内分散 $\sigma_W^2$ で評価します。これを最小化することで、各集団をコンパクトにまとめます。

$$\sigma_W^2 = \sum_{k=1}^2 \sum_{i \in I_k} (z_i - \bar{z}_k)^2 = \mathbf{w}^\top S_W \mathbf{w}$$

ここで、$S_W$ は射影前の空間における **群内分散行列** であり、次のように定義されます。

$$S_W = \sum_{k=1}^2 \sum_{i \in I_k} (\mathbf{x}_i - \bar{\mathbf{x}}_k)(\mathbf{x}_i - \bar{\mathbf{x}}_k)^\top$$

<br>

#### **(3) 最適化の指標：フィッシャー比 $J(\mathbf{w})$ と一般化固有値問題**

「中心の距離（最大化）」と「内のバラツキ（最小化）」の比を **フィッシャー比** と呼び、これを最大化することがゴールです。元のデータのバラツキを表す **散布行列 (Scatter Matrix)** $S_W, S_B$ を用いると、以下のようにまとめられます。

$$J(\mathbf{w}) = \frac{\sigma_B^2}{\sigma_W^2} = \frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_W \mathbf{w}}$$

この $J(\mathbf{w})$ を最大化する $\mathbf{w}$ は、以下の **一般化固有値問題** を解くことに帰着します。

$$S_W^{-1} S_B \mathbf{w} = \lambda \mathbf{w}$$

この最大固有値に対応する固有ベクトル $\mathbf{w}$ が、データを最も鮮明に判別できる軸となります。

### **mva_10-A**

✅ **目的**: 具体的な行列 $S_W$ と $S_B$ を用いて固有値問題を解き、数理的に導かれた「最適な軸」が、データの視覚的な分離と一致することを体験します。

#### 🖊 **数理理解：固有値問題の導出ステップ**

以下の具体的な数値例を用いて、最適な判別軸 $\mathbf{w}$ を導出します。

**設定値**：

- クラス1の平均 $\bar{\mathbf{x}}_1$ とクラス2の平均 $\bar{\mathbf{x}}_2$ の差が以下とします。
$$\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2 = \begin{pmatrix} 4 \\ 4 \end{pmatrix}$$
このとき、群間分散行列 $S_B$ は次のように定義されます。
$$S_B = (\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)(\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)^\top = \begin{pmatrix} 4 \\ 4 \end{pmatrix} \begin{pmatrix} 4 & 4 \end{pmatrix} = \begin{pmatrix} 16 & 16 \\ 16 & 16 \end{pmatrix}$$

- 群内分散行列を $S_W = \begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$ とします。

<br>

**Step 1: 逆行列 $S_W^{-1}$ の計算**

行列式 $|S_W| = (2 \times 2) - (1 \times 1) = 3$ より、

$$S_W^{-1} = \frac{1}{3} \begin{pmatrix} 2 & -1 \\ -1 & 2 \end{pmatrix}$$

<br>

**Step 2: 行列 $A = S_W^{-1} S_B$ の合成**

$$A = \frac{1}{3} \begin{pmatrix} 2 & -1 \\ -1 & 2 \end{pmatrix} \begin{pmatrix} 16 & 16 \\ 16 & 16 \end{pmatrix} = \begin{pmatrix} 16/3 & 16/3 \\ 16/3 & 16/3 \end{pmatrix}$$

<br>

**Step 3: 固有値 $\lambda$ と固有ベクトル $\mathbf{w}$ の導出**

固有方程式 $|A - \lambda I| = 0$ を解くと：

$$\begin{vmatrix} 16/3 - \lambda & 16/3 \\ 16/3 & 16/3 - \lambda \end{vmatrix} = 0 \implies (16/3 - \lambda)^2 - (16/3)^2 = 0 \implies \lambda \left( \lambda - \frac{32}{3} \right) = 0$$

最大固有値 $\lambda = 32/3$ に対する固有ベクトル $\mathbf{w}$ は、$(A - \lambda I)\mathbf{w} = \mathbf{0}$ より：

$$\begin{pmatrix} -16/3 & 16/3 \\ 16/3 & -16/3 \end{pmatrix} \begin{pmatrix} w_1 \\ w_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix} \implies w_1 = w_2$$

したがって、最適な方向は $\mathbf{w}=[1, 1]^\top$、つまり $45^\circ$ の方向であることがわかります。

#### **具体例で確認してみよう！**

以下のコードを実行して、数理的に導かれた「最適な軸」が、データの視覚的な分離と一致することを体験しましょう。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive_output, FloatSlider, VBox, HTML
from IPython.display import display

# ---------------------------------------------------------
# 1. 数値例 [S_W=(2,1,1,2)] に基づきつつ、重心を離してデータ生成
# ---------------------------------------------------------
def generate_step_by_step_data():
    np.random.seed(42)
    n_samples = 200

    # 手計算の S_W に対応する共分散構造 (サンプル数で調整)
    cov = np.array([[2, 1], [1, 2]]) * 0.4

    # 重心の距離を広げる (d = [4, 4] に設定)
    # 方向は [1, 1] のままなので、最適な射影角 45度は変わりません
    mean0 = np.array([-2.0, -2.0])
    mean1 = np.array([2.0, 2.0])

    X0 = np.random.multivariate_normal(mean0, cov, n_samples)
    X1 = np.random.multivariate_normal(mean1, cov, n_samples)

    X = np.vstack([X0, X1])
    y = np.hstack([np.zeros(n_samples), np.ones(n_samples)])
    return X, y, mean0, mean1

X, y, m0, m1 = generate_step_by_step_data()

# ---------------------------------------------------------
# 2. 理論解の定義 (手計算の結果: 45度)
# ---------------------------------------------------------
theoretical_theta = 45.0

# ---------------------------------------------------------
# 3. 描画関数
# ---------------------------------------------------------
def update_plot(theta_deg):
    theta = np.radians(theta_deg)
    v = np.array([np.cos(theta), np.sin(theta)])
    projected_X = np.dot(X, v)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
    colors = ['#3b82f6', '#ef4444'] # Blue and Red

    # --- 左側: 2次元散布図 ---
    for label, color in zip([0, 1], colors):
        ax1.scatter(X[y == label, 0], X[y == label, 1], alpha=0.5, c=color, label=f'Class {label}')

    line_len = 8
    # ユーザーが操作する軸
    ax1.plot([-v[0]*line_len, v[0]*line_len], [-v[1]*line_len, v[1]*line_len],
             'k--', linewidth=1.5, label=f'Selected Axis ({theta_deg:.1f}°)')

    # 理論解 (45度) をガイドとして表示
    v_opt = np.array([np.cos(np.radians(theoretical_theta)), np.sin(np.radians(theoretical_theta))])
    ax1.plot([-v_opt[0]*line_len, v_opt[0]*line_len], [-v_opt[1]*line_len, v_opt[1]*line_len],
             'green', linewidth=4, alpha=0.2, label='Math Solution (45°)')

    # 描画範囲をデータに合わせて調整
    ax1.set_xlim(-8, 8); ax1.set_ylim(-8, 8); ax1.set_aspect('equal')
    ax1.set_title("2D Data Space")
    ax1.legend(loc='upper left'); ax1.grid(True, alpha=0.3)

    # --- 右側: 1次元投影（ヒストグラム） ---
    for label, color in zip([0, 1], colors):
        sns.histplot(projected_X[y == label], kde=True, ax=ax2, color=color,
                     element="step", alpha=0.3, label=f'Class {label}')

    ax2.set_title(f"Projected Distribution (Projection Angle: {theta_deg:.1f}°)")
    ax2.set_xlabel("Discriminant Score (z)")
    ax2.legend()

    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------
# 4. インタラクティブUIの構築
# ---------------------------------------------------------
angle_slider = FloatSlider(
    min=0, max=180, step=1.0, value=0,
    description='Angle [deg]',
    continuous_update=True,
    layout={'width': '500px'}
)

interactive_ui = interactive_output(update_plot, {'theta_deg': angle_slider})
display(HTML("""
<div style="background-color: #f0f7ff; padding: 15px; border-radius: 10px; margin-bottom: 10px;">
    スライダーを動かして 45° 付近でヒストグラムが綺麗に分かれることを確認してください。
</div>
"""))
display(VBox([angle_slider, interactive_ui]))
```

### **mva_10-B**
✅  **目的**: Pythonのライブラリ（``scikit-learn``）を用いて線形判別分析（LDA）を実装し、決定境界がデータの分布（分散）をどのように捉えているかを視覚的に理解します。

#### 💻 **【実装・応用】 ライブラリによる線形判別分析**

実務における判別分析は、``scikit-learn`` ライブラリの ``LinearDiscriminantAnalysis`` クラスを用いることで、複雑な行列計算を意識せずに実行できます。
詳細な仕様については、以下の公式ドキュメントを参照してください。

- ``sklearn`` 公式ドキュメント: ``LinearDiscriminantAnalysis``


#### **具体例で確認してみよう！**

演習Aと類似したデータを用い、``LinearDiscriminantAnalysis`` を使って決定境界（境界線）を描画します。
特に、以前学習した ロジスティック回帰 (Logistic Regression) と比較したとき、境界線の「傾き」や「位置」がどのように変わるかに注目してください。

- **線形判別分析 (LDA)**: クラス内の分散（散らばり具合）を考慮して、最も分離が良い軸を探します。
- **ロジスティック回帰**: データの境界付近にある点（尤度）を重視して、クラスを分ける確率の境界を探します。

データが正規分布に近い場合、LDAは非常に強力な判別性能を発揮します。


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_blobs

# 1. データの生成 (演習Aの設定を再現)
def generate_comparison_data():
    np.random.seed(42)
    n_samples = 200

    # 手計算の S_W = [[2, 1], [1, 2]] に対応する共分散
    cov = np.array([[2, 1], [1, 2]]) * 0.4

    # 重心の差 d = [4, 4] を再現
    mean0 = np.array([-2.0, -2.0]) # Class 2
    mean1 = np.array([2.0, 2.0])   # Class 1

    X0 = np.random.multivariate_normal(mean0, cov, n_samples)
    X1 = np.random.multivariate_normal(mean1, cov, n_samples)

    X = np.vstack([X0, X1])
    y = np.hstack([np.zeros(n_samples), np.ones(n_samples)])
    return X, y

X, y = generate_comparison_data()

# 2. 各モデルの学習
# 線形判別分析 (LDA)
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)

# ロジスティック回帰 (Logistic Regression)
log_reg = LogisticRegression()
log_reg.fit(X, y)

# 3. 決定境界を描画するためのグリッド作成
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# 4. 可視化
plt.figure(figsize=(12, 6))

# --- 左側: 線形判別分析 (LDA) ---
plt.subplot(1, 2, 1)
Z_lda = lda.predict(np.c_[xx.ravel(), yy.ravel()])
Z_lda = Z_lda.reshape(xx.shape)
plt.contourf(xx, yy, Z_lda, alpha=0.2, cmap='RdBu_r')
plt.scatter(X[y==0, 0], X[y==0, 1], c='blue', alpha=0.5, label='Class 2', edgecolors='k')
plt.scatter(X[y==1, 0], X[y==1, 1], c='red', alpha=0.5, label='Class 1', edgecolors='k')

# 射影軸 (w) の方向を表示 (規格化して長さを一定に調整)
w = lda.coef_[0]
w_unit = w / np.linalg.norm(w) # 規格化 (Unit vector)
plt.quiver(0, 0, w_unit[0]*3, w_unit[1]*3, angles='xy', scale_units='xy', scale=1, color='green',
           label='LDA Weight Vector (normalized)', width=0.015)

plt.title("Linear Discriminant Analysis (LDA)")
plt.xlabel("$x_1$"); plt.ylabel("$x_2$")
plt.legend()
plt.grid(True, alpha=0.3)

# --- 右側: ロジスティック回帰 ---
plt.subplot(1, 2, 2)
Z_log = log_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z_log = Z_log.reshape(xx.shape)
plt.contourf(xx, yy, Z_log, alpha=0.2, cmap='RdBu_r')
plt.scatter(X[y==0, 0], X[y==0, 1], c='blue', alpha=0.5, label='Class 2', edgecolors='k')
plt.scatter(X[y==1, 0], X[y==1, 1], c='red', alpha=0.5, label='Class 1', edgecolors='k')

plt.title("Logistic Regression")
plt.xlabel("$x_1$"); plt.ylabel("$x_2$")
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 5. 結果の数値確認
print("-" * 50)
print("【LDA 算出結果】")
print(f"重みベクトル w (規格化済み): {w_unit}")
# 方向を角度に変換
angle_lda = np.degrees(np.arctan2(w_unit[1], w_unit[0]))
print(f"導き出された境界線の法線角度: {angle_lda:.2f}°")
print("-" * 50)
```
