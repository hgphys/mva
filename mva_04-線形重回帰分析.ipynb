## **要点まとめ**

**Part Ⅱ：誤差の最小化**

データ行列の中の 「特定の列（目的変数）」 を、「その他の列（説明変数）」 を使って数理的に予測することを目指します。今回は、扱うデータがすべて 量的データ（数値） である場合を対象とし、予測値と実測値のズレ（誤差エネルギー）を最小化する基本的な手法、**線形重回帰分析（Multiple Linear Regression）** を学びます。

### **1. 重回帰分析とは**

データを行列（表）の形式で整理すると、以下のようになります。各サンプル（行）について、予測したい値 $y$ と、その予測に使う情報 $x$ が対になっています。


| | 目的変数| 説明変数1 | ... | 説明変数d |
|---|:---:|:---:|:---:|:---:|
| **サンプル1** | $y_{1}$ | $x_{11}$  | ... | $x_{1d}$ |
| **サンプル2** | $y_{2}$ | $x_{21}$ | ... | $x_{2d}$ |
| ... | ... | ... | ... | ... |
| **サンプルn** | $y_{n}$ | $x_{n1}$ | ... | $x_{nd}$ |



- **目的変数 (Target / Response variable)**: $y$
  - 予測したい変数（例：売上、家賃、試験の点数）

- **説明変数 (Features / Explanatory variables)**: $x_1, \dots, x_d$
  - 予測の手がかりとなる変数（例：広さ、駅からの距離、勉強時間）

#### **数理モデル**

ある変数 $y$ （目的変数）の変化を、複数の変数 $x_1, x_2, \dots, x_d$ （説明変数）の線形結合（足し合わせ）で説明・予測するモデルです。個々のデータサンプル $i$ について、以下の関係を仮定します。

$$y_i = w_0 + w_1 x_{i1} + w_2 x_{i2} + \dots + w_d x_{id} + \epsilon_i$$

ここで、$w$ は 偏回帰係数（重み）、$\epsilon$ は 誤差（ノイズ） です。
これをすべてのデータ（$n$ 個）についてまとめると、行列形式でシンプルに表現できます。

$$\mathbf{y} = \mathbf{X}\mathbf{w} + \mathbf{e}$$
- $\mathbf{y}$: 目的変数ベクトル ($n \times 1$)
- $\mathbf{X}$: デザイン行列 ($n \times (d+1)$)。先頭列にバイアス項のための $1$ を加える。
- $\mathbf{w}$: パラメータベクトル ($(d+1) \times 1$)
- $\mathbf{e}$: 誤差ベクトル ($n \times 1$)

<br>

### **2. 最小二乗法 (OLS) と目的関数**

重回帰分析における 数理最適化問題 を定式化します。
予測値と実測値のズレである誤差ベクトル $\mathbf{e}$ の大きさ（ノルム）を評価する 目的関数 $J(\mathbf{w})$ を定義し、これを最小化するパラメータ $\mathbf{w}$ を探索します。

- **目的関数** $J(\mathbf{w})$: 残差平方和（Residual Sum of Squares, RSS）
誤差ベクトル $\mathbf{e}$ のL2ノルムの二乗として定義されます。
$$J(\mathbf{w}) = \|\mathbf{e}\|^2 = \sum_{i=1}^n (y_i - \hat{y_i})^2 = \|\mathbf{y} - \mathbf{X}\mathbf{w}\|^2$$
- **最適化**: $\nabla J(\mathbf{w}) = 0$ となる $\mathbf{w}$ を求める

この目的関数 $J$ を $\mathbf{w}$ で微分して $0$ と置くことで、以下の 正規方程式 (Normal Equation) が導かれます。

$$\mathbf{X}^\top \mathbf{X} \mathbf{w} = \mathbf{X}^\top \mathbf{y}$$

これを $\mathbf{w}$ について解くと、目的関数を最小化する最適な重み $\mathbf{\hat{w}}$ が一発で求まります。

$$\mathbf{\hat{w}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}$$

<br>

### **3. 決定係数 ($R^2$)**

構築したモデルがどのくらいデータをうまく説明できているかを表す指標として、決定係数（$R^2$） を使います。
$1$ に近いほど当てはまりが良く、$0$ に近いほど説明力が低いことを示します。

数式上は、「全変動（データのばらつき）」に対する「残差変動（モデルが説明しきれなかったばらつき）」の比率を用いて計算されます。これはベクトルのノルムを用いても表現できます。

$$R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y_i})^2}{\sum_{i=1}^n (y_i - \bar{y})^2} = 1 - \frac{\|\mathbf{y} - \mathbf{\hat{y}}\|^2}{\|\mathbf{y} - \mathbf{\bar{y}}\|^2}$$

- $\mathbf{\hat{y}} = \mathbf{X}\mathbf{\hat{w}}$ : 予測値ベクトル
- $\mathbf{\bar{y}} = \bar{y}\mathbf{1}$ : 全要素が平均値 $\bar{y}$ であるベクトル

## **Python 演習課題**

### **mva_04-A**
✅ **目的**: 最小二乗法における最適解（偏回帰係数）が、数学的にどのように導かれるか（正規方程式）を理解しましょう。また、NumPyを用いた行列演算でこれを実装し、3次元可視化を通じて「予測平面」と「誤差」の幾何学的な関係を確認しましょう。

#### 🖊 **【数理理解】正規方程式の導出と計算**

目的関数 $J(\mathbf{w})$ を最小化するための条件 $\nabla J(\mathbf{w}) = \mathbf{0}$ を求めます。
行列微分を用いて式変形を行います。

1. 目的関数の展開:
$$\begin{aligned}
J(\mathbf{w}) &= \|\mathbf{y} - \mathbf{X}\mathbf{w}\|^2 \\
&= (\mathbf{y} - \mathbf{X}\mathbf{w})^\top (\mathbf{y} - \mathbf{X}\mathbf{w}) \\
&= \mathbf{y}^\top \mathbf{y} - \mathbf{y}^\top \mathbf{X}\mathbf{w} - \mathbf{w}^\top \mathbf{X}^\top \mathbf{y} + \mathbf{w}^\top \mathbf{X}^\top \mathbf{X}\mathbf{w} \\
&= \mathbf{y}^\top \mathbf{y} - 2\mathbf{w}^\top \mathbf{X}^\top \mathbf{y} + \mathbf{w}^\top \mathbf{X}^\top \mathbf{X}\mathbf{w}
\end{aligned}$$
（スカラー値の転置は自身と等しいため、交差項をまとめました）

2. $\mathbf{w}$ で偏微分して 0 と置く:
$$\frac{\partial J(\mathbf{w})}{\partial \mathbf{w}} = - 2\mathbf{X}^\top \mathbf{y} + 2\mathbf{X}^\top \mathbf{X}\mathbf{w} = \mathbf{0}$$
行列やベクトルの微分は複雑に見えますが、高校数学の「1変数の微分」とアナロジー（類似性）で考えると理解しやすくなります。
 | 1変数の場合 ($x$ で微分) | ベクトルの場合 ($\mathbf{w}$ で微分) | 備考 |
| ----- | ----- | ----- |
| $(ax)' = a$ | $\frac{\partial}{\partial \mathbf{w}}(\mathbf{a}^\top \mathbf{w}) = \mathbf{a}$ | 1次の項の係数が残る |
| $(ax^2)' = 2ax$ | $\frac{\partial}{\partial \mathbf{w}}(\mathbf{w}^\top \mathbf{A} \mathbf{w}) = 2\mathbf{A}\mathbf{w}$ | 2次の項は係数が前に出て次数が下がる<br>(※$\mathbf{A}$が対称行列の場合) |
    - $-2\mathbf{w}^\top (\mathbf{X}^\top \mathbf{y})$ は $\mathbf{w}$ についての1次の項なので、微分すると係数部分の $-2\mathbf{X}^\top \mathbf{y}$ が残ります。
    - $\mathbf{w}^\top (\mathbf{X}^\top \mathbf{X}) \mathbf{w}$ は $\mathbf{w}$ についての2次の項で、$\mathbf{X}^\top \mathbf{X}$ は対称行列なので、微分すると$2\mathbf{X}^\top \mathbf{X} \mathbf{w}$ になります。

3. 正規方程式:
$$\mathbf{X}^\top \mathbf{X}\mathbf{w} = \mathbf{X}^\top \mathbf{y}$$

4. 解の導出:
$$\mathbf{\hat{w}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}$$


**補足：逆行列と多重共線性**

注意すべき点として、この解を求めるには、行列 $\mathbf{X}^\top \mathbf{X}$ の 逆行列 が存在する必要があります。ここで、データを中心化（平均0）した行列 $X_c$ を考えると、$\mathbf{X}^\top \mathbf{X}$ は 分散共分散行列 $\Sigma$ と以下の関係にあります。
$$\Sigma = \frac{1}{n} X_c^\top X_c \quad \Longleftrightarrow \quad X_c^\top X_c = n \Sigma$$
変数間の 相関が非常に強い（相関係数が $\pm 1$ に近い） 場合、データの分布は「潰れた楕円（直線に近い形）」になります。これは行列式が $0$ に近づくことを意味し、逆行列 $(\mathbf{X}^\top \mathbf{X})^{-1}$ が計算できない、あるいは計算結果が不安定になります。この問題を **多重共線性（Multicollinearity）** と呼びます。「似たような説明変数を同時に入れてはいけない」という実務上のルールは、数理的には「逆行列を安定して求めるための条件」と考えることができます。

<br>

#### **具体例で確認してみよう！**

「店舗の広さ」と「駅からの距離」から「1日の売上」を予測する問題を考えます。
- $y$: 1日の売上 (万円)
- $x_1$: 店舗の広さ ($\mathrm{m}^2$)
- $x_2$: 駅からの距離 (分)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# データの作成
data = {
    '売上(y)': [40, 55, 60, 75, 50],
    '広さ(x1)': [30, 40, 50, 60, 70],
    '駅距離(x2)': [10, 5, 10, 5, 20]
}

df = pd.DataFrame(data)

# データの確認
print("データセット:")
display(df)
```

##### **1. NumPyによる正規方程式の解の導出**
上記の数式通りにプログラムを書いて、パラメータ $\mathbf{w}$ を求めます。

```python
# --- 1. デザイン行列 X と 目的変数ベクトル y の準備 ---
n = len(df)
# バイアス項 w0 のための「1」の列を作成
ones = np.ones((n, 1))
# 説明変数 (x1, x2) を抽出
X_features = df[['広さ(x1)', '駅距離(x2)']].values

# 水平方向に結合してデザイン行列 X を作成: [1, x1, x2]
X = np.hstack((ones, X_features))
y = df['売上(y)'].values

print(f"デザイン行列 X の形状: {X.shape}")
print(X)

# --- 2. 正規方程式を解く ---
# w = (X^T X)^-1 X^T y

# Xの転置行列
XT = X.T

# X^T X (分散共分散行列に比例する行列)
XTX = np.dot(XT, X)

# 逆行列 (X^T X)^-1
# 注意: 実務では逆行列計算の安定性のため np.linalg.solve を使うことが多いですが、ここでは定義通り計算します
XTX_inv = np.linalg.inv(XTX)

# 最適な重み w を計算
w_hat = np.dot(np.dot(XTX_inv, XT), y)

print("\n--- 推定結果 (NumPy) ---")
print(f"切片 (w0): {w_hat[0]:.4f}")
print(f"広さの係数 (w1): {w_hat[1]:.4f}")
print(f"駅距離の係数 (w2): {w_hat[2]:.4f}")

# 数式モデルの表示
print(f"\n予測モデル: y = {w_hat[0]:.2f} + {w_hat[1]:.2f} * x1 + {w_hat[2]:.2f} * x2")
```

##### **2. 結果の可視化：予測平面と誤差の物理的解釈**
求めたパラメータ $\mathbf{w}$ がどのような平面（モデル）を作っているか、そして「誤差の最小化」が幾何学的にどう見えるかを3次元プロットで確認します。

```python
import plotly.graph_objects as go

# 1. 予測平面用のメッシュグリッドを作成
x1_range = np.linspace(df['広さ(x1)'].min(), df['広さ(x1)'].max(), 10)
x2_range = np.linspace(df['駅距離(x2)'].min(), df['駅距離(x2)'].max(), 10)
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)

# メッシュ上の予測値 y_hat を計算: y = w0 + w1*x1 + w2*x2
y_mesh = w_hat[0] + w_hat[1] * x1_mesh + w_hat[2] * x2_mesh

# 2. データ点ごとの予測値（残差ライン用）
y_pred_points = w_hat[0] + w_hat[1] * df['広さ(x1)'] + w_hat[2] * df['駅距離(x2)']

# 3. 3次元プロットの作成
fig = go.Figure()

# (a) データ点のプロット（実測値）
fig.add_trace(go.Scatter3d(
    x=df['広さ(x1)'], y=df['駅距離(x2)'], z=df['売上(y)'],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='実測値 (Data Points)'
))

# (b) 予測平面のプロット
fig.add_trace(go.Surface(
    x=x1_range, y=x2_range, z=y_mesh,
    colorscale='Viridis',
    opacity=0.5,
    name='予測平面 (Regression Plane)'
))

# (c) 残差（誤差）を表す線（バネ）のプロット
# 各データ点と、それに対応する平面上の点（予測値）を線で結ぶ
for i in range(len(df)):
    fig.add_trace(go.Scatter3d(
        x=[df['広さ(x1)'][i], df['広さ(x1)'][i]],
        y=[df['駅距離(x2)'][i], df['駅距離(x2)'][i]],
        z=[df['売上(y)'][i], y_pred_points[i]],
        mode='lines',
        line=dict(color='red', width=3),
        showlegend=False  # 凡例が大量になるのを防ぐ
    ))

# レイアウト設定
fig.update_layout(
    title='線形重回帰モデルの可視化：平面と誤差',
    scene=dict(
        xaxis_title='広さ (x1)',
        yaxis_title='駅距離 (x2)',
        zaxis_title='売上 (y)'
    ),
    width=800,
    height=600,
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
```

### **mva_04-B**
✅ **目的**: ``scikit-learn`` ライブラリを用いて重回帰分析を効率的に実装し、外れ値がモデル（予測）に与える影響を視覚的に理解しましょう。

### **💻 【実装・応用】 ``scikit-learn`` による重回帰分析と外れ値の可視化**

実務では、行列計算をゼロから実装するのではなく、最適化されたライブラリを使用します。
また、最小二乗法は「誤差の二乗和」を最小化するため、外れ値（傾向から極端に外れたデータ）の影響を強く受けやすいという特性があります。

詳細な仕様については、以下の公式ドキュメントを参照してください。
- ``scikit-learn`` 公式ドキュメント: [LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

### **具体例で確認してみよう！**

演習 A で作成したデータセットに対し、ライブラリを用いてモデルを構築します。さらに、意図的に外れ値を追加し、予測平面がどのように変化するか（歪むか）を可視化して確認します。

```python
from sklearn.linear_model import LinearRegression

# モデルの構築
model = LinearRegression()

# 学習 (fit)
# scikit-learnではバイアス項(1の列)を自動追加するため、元の説明変数のみを渡します
X_train = df[['広さ(x1)', '駅距離(x2)']]
y_train = df['売上(y)']

model.fit(X_train, y_train)

# 結果の表示
print("--- 推定結果 (Scikit-learn) ---")
print(f"切片 (intercept_): {model.intercept_:.4f}")
print(f"係数 (coef_): {model.coef_}")

# NumPyでの計算結果と比較
print("\n(参考) NumPy手計算の結果:")
print(f"切片: {w_hat[0]:.4f}, 係数: {w_hat[1:]}")

# 決定係数 (R^2) の確認
r2 = model.score(X_train, y_train)
print(f"\n決定係数 (R^2): {r2:.4f}")
```

```python
# 1. 外れ値の追加
# 非常に広くて駅に近いが、売上が低い異常データを追加
outlier_df = pd.DataFrame({
    '広さ(x1)': [80],
    '駅距離(x2)': [2],
    '売上(y)': [10]  # 通常なら高いはずが、極端に低い
})

# データを結合
df_with_outlier = pd.concat([df, outlier_df], ignore_index=True)

# 2. 外れ値ありデータでの再学習
model_outlier = LinearRegression()
X_out = df_with_outlier[['広さ(x1)', '駅距離(x2)']]
y_out = df_with_outlier['売上(y)']
model_outlier.fit(X_out, y_out)

print("--- 外れ値ありモデルの結果 ---")
print(f"切片: {model_outlier.intercept_:.4f}")
print(f"係数: {model_outlier.coef_}")
print(f"決定係数 R2: {model_outlier.score(X_out, y_out):.4f}")


# --- 3. 可視化：2つの平面の比較 ---

# メッシュデータ作成（範囲を少し広げる）
x1_range_v = np.linspace(20, 90, 20)
x2_range_v = np.linspace(0, 25, 20)
x1_mesh_v, x2_mesh_v = np.meshgrid(x1_range_v, x2_range_v)

# feature namesの警告を回避するため、予測用のメッシュデータもDataFrame化
X_mesh_df = pd.DataFrame(
    np.c_[x1_mesh_v.ravel(), x2_mesh_v.ravel()],
    columns=['広さ(x1)', '駅距離(x2)']
)

# (A) 元のモデルによる予測
y_mesh_orig = model.predict(X_mesh_df).reshape(x1_mesh_v.shape)

# (B) 外れ値ありモデルによる予測
y_mesh_out = model_outlier.predict(X_mesh_df).reshape(x1_mesh_v.shape)

# (C) 線描画用：外れ値モデルによる全データ点の予測値
y_pred_outlier_points = model_outlier.predict(df_with_outlier[['広さ(x1)', '駅距離(x2)']])

fig = go.Figure()

# 1. 元の予測（青系）
fig.add_trace(go.Surface(
    x=x1_range_v, y=x2_range_v, z=y_mesh_orig,
    colorscale='Blues', opacity=0.3,
    showscale=False,
    name='元の予測平面'
))

# 2. 外れ値あり平面（赤系）
fig.add_trace(go.Surface(
    x=x1_range_v, y=x2_range_v, z=y_mesh_out,
    colorscale='Reds', opacity=0.5,
    showscale=False,
    name='外れ値あり予測平面'
))

# 3. 通常データ点（青玉）
fig.add_trace(go.Scatter3d(
    x=df['広さ(x1)'], y=df['駅距離(x2)'], z=df['売上(y)'],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='通常データ'
))

# 4. 外れ値データ点（赤玉）
fig.add_trace(go.Scatter3d(
    x=outlier_df['広さ(x1)'], y=outlier_df['駅距離(x2)'], z=outlier_df['売上(y)'],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='外れ値 (Outlier)'
))

# 5. 外れ値モデルの残差（赤線）
# 赤い予測とデータ点を結ぶ線を描画
for i in range(len(df_with_outlier)):
    fig.add_trace(go.Scatter3d(
        x=[df_with_outlier.loc[i, '広さ(x1)'], df_with_outlier.loc[i, '広さ(x1)']],
        y=[df_with_outlier.loc[i, '駅距離(x2)'], df_with_outlier.loc[i, '駅距離(x2)']],
        z=[df_with_outlier.loc[i, '売上(y)'], y_pred_outlier_points[i]],
        mode='lines',
        line=dict(color='red', width=2),
        showlegend=False
    ))

fig.update_layout(
    title='外れ値による予測の歪みと残差（赤線）',
    scene=dict(
        xaxis_title='広さ (x1)',
        yaxis_title='駅距離 (x2)',
        zaxis_title='売上 (y)'
    ),
    width=800, height=600
)

fig.show()
```
