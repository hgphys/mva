## **フィッシャー比とマハラノビス距離の関係**

フィッシャー比 $J(\mathbf{w})$ とマハラノビス距離 $D^2$ は数理的に非常に近い関係にあります。一言で言えば、**「判別分析（LDA）とは、マハラノビス距離が最大になる『視点（投影軸）』を探す作業」** と言い換えることができます。

### **1. マハラノビス距離の復習**

2つの群（クラス1とクラス2）の重心間の距離を、データの散らばり（共分散行列 $\Sigma$）を考慮して測るのがマハラノビス距離です。

$$D^2 = (\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)^\top S_W^{-1} (\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)$$

この式の構造を、その数理的な役割で分解してみましょう：
- **平均の差の項** $(\bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2)$: これは2地点間の「単純な距離（ユークリッド距離的な成分）」を決定する要素です。
- **分散の逆行列** $S_W^{-1}$: これは 1次元における「分散で割る」操作を多次元に拡張したものです。バラツキが大きい方向の距離を割り引き、バラツキが小さい方向の距離を重視して評価する標準化の役割を担っています。

このように、マハラノビス距離は「距離を分散で割って無次元化する」という構造を持っています。

### **2. フィッシャー比の構造を分解する**

次に、フィッシャー比の式を見てみましょう。

$$J(\mathbf{w}) = \frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_W \mathbf{w}}$$

2クラスの場合、$S_B = \mathbf{d}\mathbf{d}^\top$ （ただし $\mathbf{d} = \bar{\mathbf{x}}_1 - \bar{\mathbf{x}}_2$）なので、分子は以下のように書けます。

$$\mathbf{w}^\top S_B \mathbf{w} = \mathbf{w}^\top (\mathbf{d}\mathbf{d}^\top) \mathbf{w} = (\mathbf{w}^\top \mathbf{d})^2$$

つまり、フィッシャー比は以下のようになります。

$$J(\mathbf{w}) = \frac{(\mathbf{w}^\top \mathbf{d})^2}{\mathbf{w}^\top S_W \mathbf{w}}$$

**マハラノビス的解釈**：
- **分子** $(\mathbf{w}^\top \mathbf{d})^2$: 軸 $\mathbf{w}$ に投影した後の「1次元空間での平均の差」の2乗。
- **分母** $\mathbf{w}^\top S_W \mathbf{w}$: 軸 $\mathbf{w}$ に投影した後の「1次元空間での群内分散」。

つまり、フィッシャー比とは **「ある軸 $\mathbf{w}$ の上だけで計算した、1次元のマハラノビス距離」** そのものなのです。

### **3. 最適な重み $\mathbf{w}^*$ の導出とその代入**

フィッシャー比 $J(\mathbf{w})$ を最大化する最適な重み $\mathbf{w}^*$ を導出します。

#### **導出プロセス**

$J(\mathbf{w})$ を $\mathbf{w}$ で微分して $0$ と置くことで、最大化問題を解きます。
商の微分公式より、以下の条件が得られます。

$$S_B \mathbf{w} = J(\mathbf{w}) S_W \mathbf{w}$$

ここで $J(\mathbf{w})$ はスカラー値の固有値 $\lambda$ に相当するため、これは一般化固有値問題 $S_W^{-1} S_B \mathbf{w} = \lambda \mathbf{w}$ となります。
2クラスの場合、$S_B = \mathbf{d}\mathbf{d}^\top$ であるため、代入すると：

$$S_W^{-1} (\mathbf{d}\mathbf{d}^\top) \mathbf{w} = \lambda \mathbf{w}$$

ここで $\mathbf{d}^\top \mathbf{w}$ はスカラー（投影された中心の差）であるため、上式は以下のように書き換えられます。

$$( \mathbf{d}^\top \mathbf{w} ) S_W^{-1} \mathbf{d} = \lambda \mathbf{w}$$

この等式から、ベクトル $\mathbf{w}$ はベクトル $S_W^{-1} \mathbf{d}$ と平行である必要があります。したがって、最適な重みベクトル $\mathbf{w}^*$ は次のように求まります。

$$\mathbf{w}^* \propto S_W^{-1} \mathbf{d}$$

#### **結果の代入**

この最適な $\mathbf{w}^* = S_W^{-1} \mathbf{d}$ を元の $J(\mathbf{w})$ の式に代入してみると、驚くべき結果が得られます。

$$J(\mathbf{w}^*) = \frac{( (S_W^{-1} \mathbf{d})^\top \mathbf{d} )^2}{(S_W^{-1} \mathbf{d})^\top S_W (S_W^{-1} \mathbf{d})}$$

転置の性質 $(AB)^\top = B^\top A^\top$ および $S_W$ の対称性（$S_W^\top = S_W$）を利用して分母を整理すると：

$$J(\mathbf{w}^*) = \frac{(\mathbf{d}^\top S_W^{-1} \mathbf{d})^2}{\mathbf{d}^\top (S_W^{-1} S_W) S_W^{-1} \mathbf{d}} = \frac{(\mathbf{d}^\top S_W^{-1} \mathbf{d})^2}{\mathbf{d}^\top S_W^{-1} \mathbf{d}} = \mathbf{d}^\top S_W^{-1} \mathbf{d}$$

この最終結果は、多次元空間におけるマハラノビス距離 $D^2$ そのものです。

### **4. 結論：何が起きているのか？**

- **マハラノビス距離**は、全次元（多次元空間）を使って計算した「究極の分離度」です。
- **フィッシャー比**は、1次元の軸に投影した時の「簡易的な分離度」です。
- **判別分析（LDA）** は、「どの方向からデータを見れば、投影後の世界でマハラノビス距離を100%再現できるか？」という、 **情報の損失を最小限にする軸** を探しているのです。
