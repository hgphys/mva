## **要点まとめ**
多変量解析は、複数の変数が複雑に絡み合うデータから有用な構造を抽出する技術体系です。

### **1. 多変量解析の全体像：行列（表）の視点**
データを「$n$ 行 $d$ 列の表（Excelシートのような行列）」として捉えると、各手法の役割を直感的に整理できます。
| | 変数1| 変数2 | ... | 変数d |
|---|:---:|:---:|:---:|:---:|
| **サンプル1** | $x_{11}$ | $x_{12}$ | ... | $x_{1d}$ |
| **サンプル2** | $x_{21}$ | $x_{22}$ | ... | $x_{2d}$ |
| ... | ... | ... | ... | ... |
| **サンプルn** | $x_{n1}$ | $x_{n2}$ | ... | $x_{nd}$ |

#### **データの構造**
 - **行（Rows）**: サンプル、個体、事例。
 - **列（Columns）**: 変数（変量）、次元、特徴量。

#### **多変量解析手法の分類**
- **回帰分析**: ある特定の「列」を、他の「列」を使って予測する。
- **クラスタ分析**: 「行（サンプル）」を似たもの同士でまとめ、行方向を粗視化（要約）する。
- **主成分分析**: 相関のある「列（変数）」の情報を集約し、列方向を粗視化（次元圧縮）する。
- **数量化理論**: データの中身が数値（量的）かカテゴリ（質的）かに応じて、適切な計算手法を適用する。

このように、多変量解析は **行列演算（線形代数）** そのものです。

### **mva_01-A**
✅ **目的**: 身近な「表データ」を行列として捉え直すことで、統計的な処理が「
行列演算」として記述できることを理解しましょう。

#### 🖊 **【数理理解】行列積による平均の算出**

サンプル数 $n$、変数（次元）数 $d$ のデータ行列 $X$ ($n \times d$) が与えられたとき、各変数の平均値を並べた列ベクトル（$d \times 1$）である平均ベクトル $\bar{\mathbf{x}}$ は、以下の行列演算で定義されます。
$$
\bar{\mathbf{x}} = \frac{1}{n} X^{\mathrm{T}} \mathbf{1}
$$
ここで、$\mathbf{1}$ はすべての成分が $1$ の列ベクトル（$n \times 1$）です。「データを合計する」という操作は「$\mathbf{1}$ ベクトルとの内積」として表現でき、「データを列方向に合計する」という操作は「$\mathbf{1}$ ベクトルとの行列積」として表現できます。これを $n$ で割ることで平均となります。

#### **具体例で確認してみよう！**
ここでは、以下の3人の生徒（行）の2科目（列：数学、英語）のテスト結果を例にします。
| | 数学 (Math) | 英語 (English) |
| :--- | :---: | :---: |
| **Student A** | 80 | 70 |
| **Student B** | 70 | 90 |
| **Student C** | 90 | 80 |


```python
import numpy as np
# 1. データの定義（3行2列の表
data = [
        [80, 70],  # Student A: [Math, English]
        [70, 90],  # Student B
        [90, 80]   # Student C
]

# 2. 行列 X への変換と形状(Shape)の確認
X = np.array(data)
n, d = X.shape
print(f"データ行列 X の形状: {X.shape} (n={n}, d={d})")
```

```python
# 3. 平均ベクトルの算出
# (A) 行列演算による計算: (1/n) * X^T * 1
ones_vec = np.ones(n)
mean_vec_matrix = (X.T @ ones_vec) / n
print("\n平均ベクトル (行列演算):", mean_vec_matrix)

# (B) 検算: NumPy関数
print("平均ベクトル (検算)    :", np.mean(X, axis=0))
```
