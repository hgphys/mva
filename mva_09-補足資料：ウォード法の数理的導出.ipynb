## **ウォード法の数理的導出**

階層型クラスタ分析の「ウォード法 (Ward's Method)」において、クラスタ併合時のコスト（誤差平方和の増加量）が、重心間のユークリッド距離を用いて効率的に計算できることを数学的に証明します。

### **1. 証明する命題**

2つのクラスタ $C_A, C_B$ を併合して新しいクラスタ $C_{new}$ を作成するとき、目的関数（誤差平方和）の増加量 $\Delta J$ は以下の式で表される。

$$\Delta J = \frac{n_A n_B}{n_A + n_B} ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_B||^2$$

- $n_A, n_B$: 各クラスタのデータ数
- $\mathbf{\bar{x}}_A, \mathbf{\bar{x}}_B$: 各クラスタの重心（平均ベクトル）

<br>

### **2. 定義と前提**

#### **目的関数 $J(C)$ （誤差平方和）**

あるクラスタ $C$ の凝集度（分散の大きさ）を、重心からの距離の二乗和で定義します。

$$J(C) = \sum_{\mathbf{x} \in C} ||\mathbf{x} - \mathbf{\bar{x}}_C||^2$$

#### **増加量 $\Delta J$ の定義**

併合前後の目的関数の差分です。

$$\Delta J = J(C_{new}) - \left( J(C_A) + J(C_B) \right)$$

#### **新しい重心の位置**

併合後のクラスタ $C_{new}$ の重心 $\mathbf{\bar{x}}_{new}$ は、元の重心の加重平均（内分点）となります。

$$\mathbf{\bar{x}}_{new} = \frac{n_A \mathbf{\bar{x}}_A + n_B \mathbf{\bar{x}}_B}{n_A + n_B}$$

<br>

### **3. 導出プロセス**

#### **Step 1: 誤差の分解**

クラスタ $A$ に含まれる任意のデータ点 $\mathbf{x}$ について、新しい重心 $\mathbf{\bar{x}}_{new}$ からの距離の2乗を考えます。
「元の重心 $\mathbf{\bar{x}}_A$」を経由するように項を分けて展開します。

$$\begin{aligned}
||\mathbf{x} - \mathbf{\bar{x}}_{new}||^2 &= ||(\mathbf{x} - \mathbf{\bar{x}}_A) + (\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new})||^2 \\
&= ||\mathbf{x} - \mathbf{\bar{x}}_A||^2 + ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2 + 2(\mathbf{x} - \mathbf{\bar{x}}_A) \cdot (\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new})
\end{aligned}$$
<br>

#### **Step 2: 総和をとる（クロス項の消去）**

これをクラスタ $A$ に含まれるすべてのデータ点について合計します。
ここで、重心の定義より「平均からの偏差の総和はゼロ（$\sum (\mathbf{x} - \mathbf{\bar{x}}_A) = \mathbf{0}$）」になるため、第3項（クロス項）は消えます。

$$\begin{aligned}
\sum_{\mathbf{x} \in A} ||\mathbf{x} - \mathbf{\bar{x}}_{new}||^2 &= \sum_{\mathbf{x} \in A} ||\mathbf{x} - \mathbf{\bar{x}}_A||^2 + \sum_{\mathbf{x} \in A} ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2 \\
&= J(C_A) + n_A ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2
\end{aligned}$$

同様に、クラスタ $B$ についても以下の関係が成り立ちます。

$$\sum_{\mathbf{x} \in B} ||\mathbf{x} - \mathbf{\bar{x}}_{new}||^2 = J(C_B) + n_B ||\mathbf{\bar{x}}_B - \mathbf{\bar{x}}_{new}||^2$$

<br>

#### **Step 3: $\Delta J$ の計算**

併合後の全誤差 $J(C_{new})$ は、$A$ の部分と $B$ の部分の合計です。

$$\begin{aligned}
J(C_{new}) &= \left( J(C_A) + n_A ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2 \right) + \left( J(C_B) + n_B ||\mathbf{\bar{x}}_B - \mathbf{\bar{x}}_{new}||^2 \right) \\
&= \left( J(C_A) + J(C_B) \right) + \left( n_A ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2 + n_B ||\mathbf{\bar{x}}_B - \mathbf{\bar{x}}_{new}||^2 \right)
\end{aligned}$$

定義式 $\Delta J = J(C_{new}) - (J(C_A) + J(C_B))$ に代入すると、元のクラスタ内の誤差 $J(C_A), J(C_B)$ が相殺され、重心移動によるコストだけが残ります。

$$\Delta J = n_A ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2 + n_B ||\mathbf{\bar{x}}_B - \mathbf{\bar{x}}_{new}||^2$$

<br>

#### **Step 4: 重心間距離への書き換え**

重心 $\mathbf{\bar{x}}_{new}$ は $\mathbf{\bar{x}}_A$ と $\mathbf{\bar{x}}_B$ を $n_B : n_A$ に内分する点であるため、各距離は以下のように書けます。

$$||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_{new}||^2 = \left( \frac{n_B}{n_A + n_B} \right)^2 ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_B||^2$$

$$||\mathbf{\bar{x}}_B - \mathbf{\bar{x}}_{new}||^2 = \left( \frac{n_A}{n_A + n_B} \right)^2 ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_B||^2$$

これらを $\Delta J$ の式に代入します。

$$\begin{aligned}
\Delta J &= n_A \left( \frac{n_B}{n_A + n_B} \right)^2 D^2 + n_B \left( \frac{n_A}{n_A + n_B} \right)^2 D^2 \quad (\text{ただし } D = ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_B||) \\
&= \frac{n_A n_B^2 + n_B n_A^2}{(n_A + n_B)^2} D^2 \\
&= \frac{n_A n_B (n_B + n_A)}{(n_A + n_B)^2} ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_B||^2 \\
&= \frac{n_A n_B}{n_A + n_B} ||\mathbf{\bar{x}}_A - \mathbf{\bar{x}}_B||^2
\end{aligned}$$

**証明終了 (Q.E.D.)**

<br>

### **4. 終わりに**

この式により、ウォード法では **「個々のデータ点全ての座標」を参照し直す必要がなく、「現在の重心位置」と「データ数」という代表値のみで計算が可能** であることが保証されます。これがウォード法が計算効率に優れる理由です。
