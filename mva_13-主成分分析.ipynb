## 要点まとめ

**Part VI：情報の最大化**

膨大な変数（次元）を持つデータから、その本質的な情報（分散）を損なわずに少数の指標に集約する **主成分分析 (Principal Component Analysis: PCA)** を学びます。


### **1. 主成分分析の全体像：列方向の集約（次元圧縮）**

これまでの手法と比較すると、主成分分析の役割が明確になります。クラスタ分析が「行（サンプル）」をまとめるのに対し、主成分分析は **「列（変数）」を合成して新しい軸を作る** アプローチです。

|  | 変数1 | 変数2 | ... | 変数d |  | **第1主成分 (PC1)** | **第2主成分 (PC2)** |
| ----- | ----- | ----- | ----- | ----- | ----- | ----- | ----- |
| **サンプル1** | $x_{11}$ | $x_{12}$ | ... | $x_{1d}$ | $\rightarrow$ | $z_{11}$ | $z_{12}$ |
| **サンプル2** | $x_{21}$ | $x_{22}$ | ... | $x_{2d}$ | $\rightarrow$ | $z_{21}$ | $z_{22}$ |
| ... | ... | ... | ... | ... | ... | ... | ... |
| **サンプルn** | $x_{n1}$ | $x_{n2}$ | ... | $x_{nd}$ | $\rightarrow$ | $z_{n1}$ | $z_{n2}$ |

<br>

中心化された各サンプルのベクトルを $\mathbf{x}_{i,c}$ とすると、新しい軸（主成分軸） $\mathbf{w}_k$ への射影によって得られる新しい座標値（主成分得点） $z_{ik}$ は、各変数の線形結合として以下のように定義されます。

$$z_{ik} = w_{k1}x_{i1} + w_{k2}x_{i2} + \dots + w_{kd}x_{id} = \mathbf{x}_{i,c}^\top \mathbf{w}_k$$

行列形式では、主成分得点行列 $Z$ は $Z = X_c \mathbf{W}$ と記述され、元の高次元データが新しい座標系へと変換される様子を表しています。

主成分分析は、判別分析と同様にデータを低次元へ射影（Projection）する手法ですが、判別分析が「グループ間の分離」を目的とする **教師あり学習** であるのに対し、主成分分析はラベルを使わずに「データ全体のばらつき（分散）」を最大化する方向を探す **教師なし学習** であるという違いがあります。

<br>

### **2. 目的関数の定義：分散の最大化**

中心化されたデータ行列 $X_c$（各変数から平均を引いた $n \times d$ 行列）を考えます。ここで、データの変数間の関係性を示す 分散共分散行列 $\Sigma$ は以下のように定義されます。

$$\Sigma = \frac{1}{n} X_c^\top X_c$$

主成分分析の目的は、射影後のデータ $z$ の分散を最大化することです。射影後の座標におけるデータの分散 $V(z)$ は、次のように展開して整理できます。

$$V(z) = \frac{1}{n} \sum_{i=1}^n z_i^2 = \frac{1}{n} \| \mathbf{z} \|^2 = \frac{1}{n} (X_c \mathbf{w})^\top (X_c \mathbf{w}) = \mathbf{w}^\top \left( \frac{1}{n} X_c^\top X_c \right) \mathbf{w} = \mathbf{w}^\top \Sigma \mathbf{w}$$

この $V(z)$ こそが最大化すべき目的関数 $J(\mathbf{w})$ であり、単位ベクトル（$\|\mathbf{w}\|^2 = 1$）という制約の下で最適化を行います。

<br>

### **3. 固有値問題への帰着と主成分の性質**

ラグランジュの未定乗数法を用いて、制約条件 $\|\mathbf{w}\|^2 = 1$ の下で $J(\mathbf{w})$ を最大化すると、以下の 固有値問題 に帰着します。

$$\Sigma \mathbf{w} = \lambda \mathbf{w}$$

分散共分散行列 $\Sigma$ は実対称行列であるため、この方程式を解くことで最大 $d$ 個の固有値 $\lambda_k$ と、それに対応する固有ベクトル $\mathbf{w}_k$ が得られます。これらは以下の重要な性質を持ちます。

- **主成分の順序**: 固有値を大きい順に $\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_d$ と並べたとき、対応する固有ベクトル $\mathbf{w}_k$ を第 $k$ 主成分の軸と呼びます。第1主成分は、データが最も大きく散らばっている方向を示します。

- **主成分の直交性**: 異なる主成分の軸同士は互いに直交します（$\mathbf{w}_i^\top \mathbf{w}_j = 0 \quad (i \neq j)$）。これは、第2主成分以降が「前の主成分が説明しきれなかった情報の残差」のうち、最も分散が大きい方向を、既存の軸と重複しないように（独立に）探すためです。

- **情報の解像度**: 固有値 $\lambda_k$ は、その主成分軸上にデータを射影したときの分散そのものを表します（$J(\mathbf{w}_k) = \lambda_k$）。つまり、固有値が大きいほど、その軸は元のデータの情報を多く保持していることになります。


<br>

### **4. 圧縮の評価：スペクトル分解と寄与率**

主成分分析による情報の分解は、数学的には分散共分散行列 $\Sigma$ の スペクトル分解（固有値分解） として理解すると、次元圧縮の意味がより明確になります。

分散共分散行列 $\Sigma$ は、固有値 $\lambda_i$ と固有ベクトル $\mathbf{w}_i$ を用いて以下のように和の形で書き表すことができます。

$$\Sigma = \lambda_1 \mathbf{w}_1 \mathbf{w}_1^\top + \lambda_2 \mathbf{w}_2 \mathbf{w}_2^\top + \dots + \lambda_d \mathbf{w}_d \mathbf{w}_d^\top = \sum_{i=1}^d \lambda_i \mathbf{w}_i \mathbf{w}_i^\top$$

この式は、元のデータの複雑な情報（$\Sigma$）が、各主成分が持つ情報（$\lambda_i \mathbf{w}_i \mathbf{w}_i^\top$）の単純な足し算に分解できることを示しています。
ここで、もし下位の固有値（例えば $\lambda_{k+1}$ 以降）が非常に小さい値であれば、それらの項を無視しても元の情報の大部分を再現できることになります。

#### **寄与率と累積寄与率**

この「近似の精度」を定量化するために、以下の指標を用います。

- **寄与率 (Proportion of Variance)**:
第 $k$ 主成分が全体の情報のどれだけを担っているかを示す指標 $C_k$ です。
$$C_k = \frac{\lambda_k}{\sum_{j=1}^d \lambda_j}$$

- **累積寄与率 (Cumulative Proportion)**:
第 $1$ から第 $k$ 主成分までで、全体の情報の何割を説明できているかを示す指標 $R_k$ です。
$$R_k = \frac{\sum_{i=1}^k \lambda_i}{\sum_{j=1}^d \lambda_j}$$

累積寄与率 $R_k$ が 0.8 〜 0.9 (80% 〜 90%) に達する次元数 $k$ を採用するのが一般的です。これは、「元の行列 $\Sigma$ の情報の8〜9割を再現できる最小の次元数」を選ぶことに相当します。

## **Python 演習課題**

### **mva_13-A**
✅ **目的**: 具体的な分散共分散行列 $\Sigma$ を用いて固有値問題を解き、数理的に導かれた「分散最大化軸」が、データの視覚的な広がりと一致することを確認しましょう。

#### 🖊 **【数理理解】固有値問題の導出ステップ**

以下の具体的な数値例を用いて、第1主成分ベクトル $\mathbf{w}$ を導出します。

**設定値**：

ある2次元データの分散共分散行列 $\Sigma$ が以下で与えられたとします。
（各変数の分散が $5$、共分散が $3$ であるような、正の相関を持つデータです）

$$\Sigma = \begin{pmatrix} 5 & 3 \\ 3 & 5 \end{pmatrix}$$

主成分分析の目的は、投影後の分散 $J(\mathbf{w}) = \mathbf{w}^\top \Sigma \mathbf{w}$ を最大化する単位ベクトル $\mathbf{w}$ を見つけることです。

<br>

**Step 1: ラグランジュの未定乗数法による立式**

目的関数と制約条件を具体的に書き下します。
$\mathbf{w} = \begin{pmatrix} w_1 \\ w_2 \end{pmatrix}$ と置くと、分散共分散行列の成分を用いて次のように表せます。

- **目的関数**:
$$J(\mathbf{w}) = \mathbf{w}^\top \Sigma \mathbf{w}= 5w_1^2 + 6w_1w_2 + 5w_2^2$$
- **制約条件**:
$$g(\mathbf{w}) = w_1^2 + w_2^2 - 1 = 0$$

<br>
これらを用いて、ラグランジュ関数 $L$ を定義します。

$$L = 5w_1^2 + 6w_1w_2 + 5w_2^2 - \lambda(w_1^2 + w_2^2 - 1)$$

$w_1, w_2$ でそれぞれ偏微分して $0$ と置きます。

$$\begin{cases}
\frac{\partial L}{\partial w_1} = 10w_1 + 6w_2 - 2\lambda w_1 = 0 \\
\frac{\partial L}{\partial w_2} = 6w_1 + 10w_2 - 2\lambda w_2 = 0
\end{cases}$$

両辺を2で割って整理すると、以下の連立方程式が得られます。

$$\begin{cases}
5w_1 + 3w_2 = \lambda w_1 \\
3w_1 + 5w_2 = \lambda w_2
\end{cases}
\quad \implies \quad
\begin{pmatrix} 5 & 3 \\ 3 & 5 \end{pmatrix} \begin{pmatrix} w_1 \\ w_2 \end{pmatrix} = \lambda \begin{pmatrix} w_1 \\ w_2 \end{pmatrix}$$

これはまさに 固有方程式 $\Sigma \mathbf{w} = \lambda \mathbf{w}$ そのものです。

<br>

**Step 2: 固有値の導出**

固有方程式 $|\Sigma - \lambda I| = 0$ を解いて、固有値 $\lambda$ を求めます。

$$\begin{vmatrix} 5 - \lambda & 3 \\ 3 & 5 - \lambda \end{vmatrix} = (5 - \lambda)^2 - 3^2 = 0$$

$$(5 - \lambda)^2 = 9 \implies 5 - \lambda = \pm 3$$

よって、固有値は $\lambda = 8, 2$ となります。


<br>


**Step 3: 最大固有値に対応する固有ベクトル $\mathbf{w}$ の導出**

分散を最大化するのは、最大固有値 $\lambda_1 = 8$ です。対応する固有ベクトル $\mathbf{w} = \begin{pmatrix} w_1 \\ w_2 \end{pmatrix}$ を求めます。
$(\Sigma - 8 I)\mathbf{w} = \mathbf{0}$ より：

$$\begin{pmatrix} 5-8 & 3 \\ 3 & 5-8 \end{pmatrix} \begin{pmatrix} w_1 \\ w_2 \end{pmatrix} = \begin{pmatrix} -3 & 3 \\ 3 & -3 \end{pmatrix} \begin{pmatrix} w_1 \\ w_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$$

これより $-3w_1 + 3w_2 = 0$、すなわち $w_1 = w_2$ という関係が得られます。
単位ベクトルという制約 $\|\mathbf{w}\|^2 = w_1^2 + w_2^2 = 1$ を適用すると：

$$w_1^2 + w_1^2 = 1 \implies 2w_1^2 = 1 \implies w_1 = \frac{1}{\sqrt{2}}$$

したがって、求める第1主成分軸は $\mathbf{w} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix}$、角度にすると 45度の方向であることがわかります。

<br>

### **具体例で確認してみよう！**

以下のコードを実行して、数理的に導かれた「分散最大化軸（45度）」が、実際のデータの広がりと一致することを体験しましょう。
スライダーを動かして、ヒストグラムの幅（Variance）が最大になる角度を探してください。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive_output, FloatSlider, VBox, HTML
from IPython.display import display

# ---------------------------------------------------------
# 1. データの読み込み (Google スプレッドシート)
# ---------------------------------------------------------
# 共有リンクの末尾を "/export?format=csv" に変更して読み込みます
sheet_url = "https://docs.google.com/spreadsheets/d/1rw5vdE1ofr38YH3vOKfnvDfBLBuZi98vFWr9FyYfrAU/export?format=csv"

df = pd.read_csv(sheet_url)
X = df.iloc[:, :2].values

# ---------------------------------------------------------
# 2. 前処理と正解の計算
# ---------------------------------------------------------
# 中心化 (PCAの前提: 重心を原点に持ってくる)
X_c = X - np.mean(X, axis=0)

# 正解（PCAによる第1主成分）を計算しておく（描画の正解ガイド用）
# np.cov はデフォルトで不偏分散 (N-1 で割る) を計算します
cov_mat = np.cov(X_c.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
idx = eig_vals.argsort()[::-1] # 固有値の降順インデックス

v_opt = eig_vecs[:, idx][:, 0]  # 第1主成分ベクトル (最大固有値に対応)
var_opt = eig_vals[idx][0]      # 最大分散 (最大固有値)

# ★全分散 (固有値の総和) を計算: これを分母にして寄与率を出します
total_variance = np.sum(eig_vals)

# ---------------------------------------------------------
# 3. インタラクティブな描画関数の定義
# ---------------------------------------------------------
def update_plot(theta_deg):
    theta = np.radians(theta_deg)
    # ユーザーが選択した方向ベクトル v = (cosθ, sinθ)
    v = np.array([np.cos(theta), np.sin(theta)])

    # データの射影 (z = X_c . v)
    z = X_c @ v

    # この方向の分散 V(z)
    # np.cov と合わせるため、ddof=1 (不偏分散) を指定します
    current_variance = np.var(z, ddof=1)

    # --- 描画設定 ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # データの最大範囲を取得（軸のスケール固定用）
    limit = np.max(np.abs(X_c)) * 1.5

    # (左図) 2次元散布図と射影軸
    ax1.scatter(X_c[:, 0], X_c[:, 1], alpha=0.5, color='#3b82f6', label='Centered Data')

    # ユーザーが選択している軸（赤破線）
    line_len = limit * 0.9
    ax1.plot([-v[0]*line_len, v[0]*line_len], [-v[1]*line_len, v[1]*line_len],
             'r--', linewidth=2, label=f'Selected Axis ({theta_deg:.1f}°)')

    # 正解の軸（薄い緑色、ガイドとして表示）
    ax1.plot([-v_opt[0]*line_len, v_opt[0]*line_len], [-v_opt[1]*line_len, v_opt[1]*line_len],
             'green', linewidth=5, alpha=0.2, label=f'PC1 (True Max Var)')

    ax1.set_xlim(-limit, limit); ax1.set_ylim(-limit, limit)
    ax1.set_xlabel('x1'); ax1.set_ylabel('x2')
    ax1.set_title("2D Data Space\nSelect the axis that maximizes the spread!")
    ax1.legend(loc='upper left'); ax1.grid(True, alpha=0.3); ax1.set_aspect('equal')

    # (右図) 射影されたデータのヒストグラム（1次元分布）
    sns.histplot(z, kde=True, ax=ax2, color='red', bins=30, element="step", alpha=0.3)

    # ★寄与率 (Contribution Ratio) の計算: 現在の分散 / 全分散
    contribution_ratio = current_variance / total_variance * 100

    # タイトルに寄与率を表示
    ax2.set_title(f"Projected Distribution (1D)\nVariance: {current_variance:.2f} (Contribution: {contribution_ratio:.1f}%)")
    ax2.set_xlabel("Principal Component Score (z)")
    ax2.set_xlim(-limit, limit)
    ax2.set_ylim(0, len(X) * 0.3) # ヒストグラムの高さを固定

    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------
# 4. UI構築
# ---------------------------------------------------------
# 0度〜180度まで探索できるようにする
slider = FloatSlider(
    min=0, max=180, step=1.0, value=90, # 初期値はあえて正解(45度)からずらす
    description='Angle [deg]',
    continuous_update=True,
    layout={'width': '500px'}
)

ui = interactive_output(update_plot, {'theta_deg': slider})

display(HTML(f"""
<div style="background-color: #f0f7ff; padding: 15px; border-radius: 10px; margin-bottom: 10px;">
    スライダーを動かして赤い点線（射影軸）を回転させてください。<br>
    </ul>
</div>
"""))
display(VBox([slider, ui]))
```

### **mva_13-B**
✅ **目的**: Pythonのライブラリ（``scikit-learn``）を用いて主成分分析（PCA）を実装し、3次元データが2次元の「情報最大化平面」へどのように射影されるかを視覚的に理解します。

#### 💻 **【実装・応用】 ライブラリによる主成分分析**

実務における次元圧縮は、``scikit-learn`` ライブラリの ``PCA`` クラスを用いることで簡単に実行できます。
詳細な仕様については、以下の公式ドキュメントを参照してください。

- ``sklearn`` 公式ドキュメント: ``PCA`` (Principal Component Analysis)

<br>

#### **具体例で確認してみよう！**

以下のコードを実行して、ラグビーボールのような形状をした3次元データを生成し、それをPCAによって「最も情報の多い（平べったい）面」に押しつぶす様子を確認しましょう。
3Dグラフをマウスで回転させて、生成された「PC1-PC2平面」がデータの広がりをどのように捉えているか観察してください。

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from sklearn.decomposition import PCA

# ==============================================================================
# mva_13-B: 3次元PCAの可視化（Plotly版）
# ==============================================================================

# ---------------------------------------------------------
# 1. ラグビーボール型データの生成
# ---------------------------------------------------------
np.random.seed(42)
n_samples = 500

# 主成分方向の分散比を設定（形状を決める）
# PC1(長軸) : PC2(中軸) : PC3(短軸)
variances = [10.0, 3.0, 0.2]
std_devs = np.sqrt(variances)

# 1. 軸に沿った正規分布データを生成 (Z)
Z = np.random.normal(0, 1, (n_samples, 3)) * std_devs

# 2. データを回転させる（斜めに配置して相関を作る）
# 適当な回転行列を作成（オイラー角などで定義）
theta = np.radians(45) # x軸周りの回転
phi = np.radians(30)   # z軸周りの回転

Rx = np.array([
    [1, 0, 0],
    [0, np.cos(theta), -np.sin(theta)],
    [0, np.sin(theta), np.cos(theta)]
])
Rz = np.array([
    [np.cos(phi), -np.sin(phi), 0],
    [np.sin(phi), np.cos(phi), 0],
    [0, 0, 1]
])
R = Rz @ Rx # 回転行列の合成

# データを回転させて平均をずらす
X_3d = Z @ R.T + np.array([1, 1, 1])

# DataFrame化（確認用）
df_3d = pd.DataFrame(X_3d, columns=['x1', 'x2', 'x3'])

# ---------------------------------------------------------
# 2. PCAの実行
# ---------------------------------------------------------
pca = PCA(n_components=3)
pca.fit(X_3d)

# 平均ベクトルと主成分ベクトル
mean_vec = pca.mean_
components = pca.components_
explained_variance = pca.explained_variance_
explained_ratio = pca.explained_variance_ratio_

print(f"第1主成分 寄与率: {explained_ratio[0]*100:.1f}%")
print(f"第2主成分 寄与率: {explained_ratio[1]*100:.1f}%")
print(f"第3主成分 寄与率: {explained_ratio[2]*100:.1f}%")
print(f"累積寄与率 (PC1+PC2): {sum(explained_ratio[:2])*100:.1f}%")

# ---------------------------------------------------------
# 3. Plotlyによるインタラクティブ可視化
# ---------------------------------------------------------
fig = go.Figure()

# (1) データ点のプロット
fig.add_trace(go.Scatter3d(
    x=X_3d[:, 0], y=X_3d[:, 1], z=X_3d[:, 2],
    mode='markers',
    marker=dict(
        size=2,
        color='brown',
        opacity=0.6
    ),
    name='Data Points'
))

# (2) 主成分軸（ベクトル）の描画
# PC1(赤), PC2(青), PC3(緑) をすべて表示します
colors = ['red', 'blue', 'green']
labels = ['PC1', 'PC2', 'PC3']
scale_factor = 2.5 # ベクトルの表示倍率（見た目の調整用）

for i in range(3): # PC1, PC2, PC3 全てを描画
    # 固有値のルート（標準偏差）に比例した長さにする
    vec_len = np.sqrt(explained_variance[i]) * scale_factor
    vec = components[i] * vec_len

    start = mean_vec
    end = mean_vec + vec
    end_neg = mean_vec - vec # 反対方向

    # 軸の描画（中心から両側に伸ばす）
    fig.add_trace(go.Scatter3d(
        x=[end_neg[0], end[0]],
        y=[end_neg[1], end[1]],
        z=[end_neg[2], end[2]],
        mode='lines+text',
        line=dict(color=colors[i], width=6),
        text=["", labels[i]], # 端点にラベルを表示
        textposition="top center",
        name=f'{labels[i]} Axis'
    ))

# (3) PC1-PC2 平面の描画（分散最大化平面）
# グリッドを作成
grid_size = 20
# PC1方向の広がりに合わせた範囲設定
range_val = np.sqrt(explained_variance[0]) * 3
u = np.linspace(-range_val, range_val, grid_size)
v = np.linspace(-range_val, range_val, grid_size)
U, V = np.meshgrid(u, v)

# 平面上の点の座標 P = mean + u*PC1 + v*PC2
# components[0] が PC1ベクトル, components[1] が PC2ベクトル
plane_x = mean_vec[0] + U * components[0, 0] + V * components[1, 0]
plane_y = mean_vec[1] + U * components[0, 1] + V * components[1, 1]
plane_z = mean_vec[2] + U * components[0, 2] + V * components[1, 2]

fig.add_trace(go.Surface(
    x=plane_x, y=plane_y, z=plane_z,
    opacity=0.3,          # 半透明にする
    colorscale='Gray',
    showscale=False,      # カラーバーは不要
    name='PC1-PC2 Plane'
))

# レイアウト設定
fig.update_layout(
    title='PCA Visualization',
    scene=dict(
        xaxis_title='x1',
        yaxis_title='x2',
        zaxis_title='x3',
        aspectmode='data' # アスペクト比をデータに合わせる（歪ませない）
    ),
    width=900, height=700,
    margin=dict(l=0, r=0, b=0, t=50),
    legend=dict(x=0, y=0.9)
)

fig.show()
```
