## 要点まとめ

**Part VI：情報の最大化**

主成分分析（PCA）では、「量的データ」の分散を最大化することで情報を圧縮しました。今回の **数量化Ⅲ類（Quantification Theory Type III）** は、アンケート回答のような **「質的データ」** を対象に、その背後に潜む「サンプル」と「カテゴリ」の関連性を抽出する手法であり、**「対応分析（Correspondence Analysis）」** として広く知られています。


### **1. 数量化Ⅲ類とは**

アンケートのクロス集計表のような、行（サンプル）や列（カテゴリ）が非常に多い「巨大な表」を想像してください。数字が羅列されただけの表からは、全体像や傾向を読み取ることは非常に困難です。

数量化Ⅲ類（対応分析）は、この **「複雑なクロス集計表」を「シンプルな地図（散布図）」に要約し、視覚的にわかりやすく表現する** ための手法です。

具体的には、**サンプルの特徴** と **カテゴリ（項目）の特徴** を同時に数値化（スコアリング）し、**「高いスコアの人は高いスコアの項目を選んでいる（正の相関がある）」** という構造を抽出し、似たもの同士を近くに配置します。数量化Ⅲ類は、**「質的データに対するPCA」** と解釈できます。

#### **クロス集計表：データの行和と列和**

以下のような、行に回答者（サンプル）、列に選択肢（カテゴリ）を並べた表形式のデータを扱います。 ここで重要なのが、右端の **「行和 ($r_i$)」** と下端の **「列和 ($c_j$)」** です。


|  | 項目1 | 項目2 | ... | 項目d |  **行和 (Total)** |
| ----- | ----- | ----- | ----- | ----- | ----- |
| **サンプル1** | 1 | 0 | ... | 1 | $r_1$|
| **サンプル2** | 1 | 0 | ... | 1 | $r_2$|
| ... | ... | ... | ... | ... |
| **サンプルn** | 1 | 1 | ... | 0  | $r_{n}$ |
| **列和 (Total)** | $c_1$ | $c_2$ | ... | $c_d$  | $N$ (総和) |

- **行和** ($r_i$): サンプル $i$ が「いくつの項目に反応したか」（回答者のアクティブ度）。

- **列和** ($c_j$): 項目 $j$ が「何人のサンプルに選ばれたか」（項目の人気度）。


<br>


### **2. 数量化Ⅲ類の数理：相関の最大化**

数量化Ⅲ類の目的は、サンプル（行）とカテゴリ（列）にそれぞれ適切なスコア（数値）を与え、両者の相関を最大化することです。

#### **(1) データの定義と対角行列**

まず、分析の基礎となる「データ行列」と、そこから導かれる「クロス集計表（共起行列）」の関係を整理します。

**データ行列** $X$: 上表のような $n$ 行 $d$ 列の $0/1$ 行列（ダミー変数行列）。

**クロス集計表（共起行列）**: データ行列から計算される行列積 $X^\top X$。
- $X^\top X$ の $(j, k)$ 成分は、項目 $j$ と項目 $k$ の **同時出現回数（共起数）** を表します。

これらを用いて、計算に必要な重み行列（対角行列）を定義します。

- $D_r$ **(行の重み行列)**: $X$ の行和（各サンプルの回答数）を対角成分に持つ $n \times n$ 行列。
$$D_r = \begin{pmatrix} r_1 & 0 & \dots \\ 0 & r_2 & \dots \\ \vdots & \vdots & \ddots \end{pmatrix}$$

- $D_c$ **(列の重み行列)**: $X$ の列和（各項目の選択数）、つまりクロス集計表 $X^\top X$ の対角成分を並べた $d \times d$ 行列。
$$D_c = \begin{pmatrix} c_1 & 0 & \dots \\ 0 & c_2 & \dots \\ \vdots & \vdots & \ddots \end{pmatrix}$$

<br>

#### **(2) スコアベクトル**

$a_i$ と $b_j$ はどちらも「スコア」ですが、役割が異なります。

**サンプルスコア** $\mathbf{a} = [a_1, \dots, a_n]^\top$
- 「人（行）」に割り当てられる座標です。似た回答パターンを持つ人同士は、近い値（近い座標）になります。

**カテゴリスコア** $\mathbf{b} = [b_1, \dots, b_d]^\top$
- 「項目（列）」に割り当てられる座標です。同じような人に選ばれやすい項目同士は、近い値（近い座標）になります。

<br>

#### **(3) 目的関数の設定**

数量化Ⅲ類のタスクは、データ行列の中に埋もれている「似たもの同士（人と項目）の結びつき」を数値化して取り出すことです。
そこで、「高いスコアの人は高いスコアの項目を選んでいる（正の相関がある）」状態を目指して、以下の目的関数（相関係数）$J$ を最大化します。

$$J = \frac{\sum_i \sum_j x_{ij} a_i b_j}{\sqrt{\sum_i r_i a_i^2} \sqrt{\sum_j c_j b_j^2}}= \frac{\mathbf{a}^\top X \mathbf{b}}{\sqrt{\mathbf{a}^\top D_r \mathbf{a}} \sqrt{\mathbf{b}^\top D_c \mathbf{b}}}$$

**【式の意味】**

**分子**： $\mathbf{a}^\top X \mathbf{b}$
- **「マッチングの合計得点」** です。「反応あり（$x_{ij}=1$）」のペアのスコア積 ($a_i \times b_j$) の合計です。似たスコア同士が結びつけば値が大きくなります。
- **クロス集計表との関係**: もし $X$ が $0/1$ ではなく、集計済みの「頻度データ（一般的なクロス集計表）」だったとしても、この式は有効です。その場合、「（頻度）×（行スコア）×（列スコア）」の総和となり、表全体の「情報の強さ（相関）」を数値化したものと解釈できます。

**分母**： $\sqrt{\mathbf{a}^\top D_r \mathbf{a}} \sqrt{\mathbf{b}^\top D_c \mathbf{b}}$
-  **「分散（スケーリング）による調整」** です。ここで $D_r, D_c$ が登場します。「回答数が多い人 ($r_i$)」や「人気のある項目 ($c_j$)」は、合計点への影響力が大きくなりすぎるため、この項で重みを調整（正規化）しています。

<br>

#### **(4) 固有値問題への帰着**

この目的関数 $J$ の最大化は、ラグランジュの未定乗数法を用いることで、一般化固有値問題 $X^\top D_r^{-1} X \mathbf{b} = \lambda D_c \mathbf{b}$ に帰着します。

さらに、この式の両辺に左から $D_c^{-1}$ を掛けることで、より直感的な**標準固有値問題**の形に変形できます。また、サンプルスコア $\mathbf{a}$ についても同様の議論が成り立ちます。これら2つの式を並べると、美しい対称性（双対性）が見えてきます。

$$\begin{aligned}
(D_c^{-1} X^\top D_r^{-1} X) \mathbf{b} &= \lambda \mathbf{b} \quad (\text{カテゴリスコア}) \\
(D_r^{-1} X D_c^{-1} X^\top) \mathbf{a} &= \lambda \mathbf{a} \quad (\text{サンプルスコア})
\end{aligned}$$

<br>

##### **【数式の意味：確率遷移行列の固有値問題】**

この変形後の式は、数量化Ⅲ類の本質を見事に表しています。カッコ内の行列 $P = D_c^{-1} X^\top D_r^{-1} X$ は、以下のように分解して解釈できます。

**行の確率化（Row Profiles）**: $D_r^{-1} X$
- 各サンプル（行）の回答を、合計が1になる「確率分布」に変換します。これにより、回答数の多い・少ないの影響が取り除かれます。

**列の確率化（Column Profiles）**: $D_c^{-1} X^\top$
- 同様に、各カテゴリ（列）の選択分布を確率化します。

つまり、行列 $P$ は「項目 $\to$ 人 $\to$ 項目」という **確率的な推移を表す行列** となります。 この行列の固有値問題を解くことは、**「確率の連鎖の中で変化しない安定した構造（軸）」** を抽出することに他なりません。同様に、行列 $Q = D_r^{-1} X^\top D_c^{-1} X$ は、「人 $\to$ 項目 $\to$ 人」という推移サイクルを表します。

上記の2つの式で重要なのは、**共通の固有値** $\lambda$ が現れるという点です。相関（の2乗）である $\lambda$ は、「人と項目の結びつきの強さ」 そのものを表すため、どちらの視点から計算しても結果は一致します。この性質（**双対性**）があるため、人と項目を同じ基準（同じ土俵）で分析・配置することができるのです。

**【固有値 $\lambda$ と固有ベクトル $\mathbf{b}$ 】**

固有値 $\lambda$ の意味:
相関係数 $J$ の2乗 ($J^2$) に対応します。$\lambda$ が大きいほど、その軸におけるサンプルとカテゴリの相関（結びつき）が強いことを示します。

固有ベクトル $\mathbf{b}$ の意味: 相関を最大化する「最適なカテゴリスコア」（第1軸）となります。

<br>

### **3. 結果の解釈：相互平均法の視点**

数量化Ⅲ類で得られるスコアには、以下の直感的な性質（相互平均法：Reciprocal Averaging）があります。

- **サンプルのスコア**: そのサンプルが選択したカテゴリのスコアの平均（に近い値）になる。
- **カテゴリのスコア**: そのカテゴリを選択したサンプルのスコアの平均（に近い値）になる。

この性質により、 **「似た回答をした人は近くに」「よく一緒に選ばれるカテゴリは近くに」**、そして **「そのカテゴリを選びがちな人は、そのカテゴリの近くに」** 配置されます。これを **バイプロット** と呼び、データの構造を多角的に把握するのに適しています。

## **Python 演習課題**

### **mva_14-A**

✅ **目的**: 具体的なデータ行列 $X$ を用いて、数量化Ⅲ類（対応分析）の計算プロセスである「確率行列の固有値問題」を解き、数理的に導かれたスコアがどのような意味を持つかを確認しましょう。

#### **🖊 【数理理解】固有値問題の導出ステップ**

以下の極めてシンプルな数値例を用いて、相関を最大化するカテゴリスコア $\mathbf{b}$ を導出します。

**設定値**：

ある2人のユーザーが、2つのアイテム（A, B）に対して以下の反応を示したとします。
- ユーザー1: アイテムA と アイテムB の両方を選んだ。
- ユーザー2: アイテムA だけを選んだ。

これをデータ行列 $X$（ダミー変数行列）で表すと以下のようになります。

$$X = \begin{pmatrix} 1 & 1 \\ 1 & 0 \end{pmatrix}$$

数量化Ⅲ類の目的は、このデータから導かれる確率遷移行列 $P = D_c^{-1} X^\top D_r^{-1} X$ の固有値問題を解くことです。

<br>

**Step 1: 偏微分による固有方程式の導出**

行列計算に入る前に、この具体例で「なぜ固有値問題を解くことになるのか」を、相関の最大化（偏微分）から確認してみましょう。

1. **目的関数の設定**
まず、数量化Ⅲ類の目的関数（最大化したい相関係数 $J$）は以下の式で表されます。
$$J = \frac{\mathbf{a}^\top X \mathbf{b}}{\sqrt{\mathbf{a}^\top D_r \mathbf{a}} \sqrt{\mathbf{b}^\top D_c \mathbf{b}}}$$
この最大化問題は、「分母（分散）を $1$ に固定した制約条件の下で、分子（共分散 $\mathbf{a}^\top X \mathbf{b}$）を最大化する」問題と等価です。
ラグランジュの未定乗数法を用いると、以下の関数 $L$ を最大化することになります。
$$L = \mathbf{a}^\top X \mathbf{b} - \frac{\lambda}{2} (\mathbf{a}^\top D_r \mathbf{a} - 1) - \frac{\lambda}{2} (\mathbf{b}^\top D_c \mathbf{b} - 1)$$
今回の具体的な数値（$X, D_r, D_c$）を代入して成分で書き下すと、以下のようになります。
（係数の $2$ や $1$ は、それぞれの行和・列和に対応します）
$$L =(a_1 b_1 + a_1 b_2 + a_2 b_1)- \frac{\lambda}{2} (2a_1^2 + a_2^2 - 1) - \frac{\lambda}{2} (2b_1^2 + b_2^2 - 1) $$
<br>

2. **偏微分して 0 と置く**
各変数で偏微分すると、以下の「相互平均（Reciprocal Averaging）」の関係式が導かれます。
- 人のスコア ($\partial L / \partial a_i = 0$): 「選んだ項目の平均」になる
$$\lambda a_1 = \frac{1}{2}(b_1 + b_2) , \quad \lambda a_2 = \frac{1}{1}(b_1)$$
- 項目のスコア ($\partial L / \partial b_j = 0$): 「選んだ人の平均」になる
$$\lambda b_1 = \frac{1}{2}(a_1 + a_2) , \quad \lambda b_2 = \frac{1}{1}(a_1)$$
<br>

3. **連立方程式を解く ($a$ を消去)**
上の式を下の式に代入して、$a$ を消去し $\mathbf{b}$ だけの式にします。
$$\lambda (\lambda b_1) = \frac{1}{2}(\lambda a_1 + \lambda a_2) = \frac{1}{2} \left( \frac{b_1 + b_2}{2} + b_1 \right) = 0.75 b_1 + 0.25 b_2$$
$$\lambda (\lambda b_2) = \frac{1}{1}(\lambda a_1) = \frac{b_1 + b_2}{2} = 0.5 b_1 + 0.5 b_2$$
<br>

これを行列形式で書くと：
$$\begin{pmatrix} 0.75 & 0.25 \\ 0.5 & 0.5 \end{pmatrix} \begin{pmatrix} b_1 \\ b_2 \end{pmatrix} = \lambda^2 \begin{pmatrix} b_1 \\ b_2 \end{pmatrix}$$

となり、「ある行列の固有値問題」に帰着することが確認できました！
次からのステップでは、この行列を定義通りに計算して導出します。

<br>

**Step 2: 重み行列 $D_r, D_c$ の定義**

まず、行和（各ユーザーの回答数）と列和（各アイテムの選択数）を計算し、対角行列を作ります。

行和ベクトル $r = [2, 1]$ $\implies D_r = \begin{pmatrix} 2 & 0 \\ 0 & 1 \end{pmatrix}$

列和ベクトル $c = [2, 1]$ $\implies D_c = \begin{pmatrix} 2 & 0 \\ 0 & 1 \end{pmatrix}$

これらの逆行列は以下の通りです。

$$D_r^{-1} = \begin{pmatrix} 0.5 & 0 \\ 0 & 1 \end{pmatrix}, \quad D_c^{-1} = \begin{pmatrix} 0.5 & 0 \\ 0 & 1 \end{pmatrix}$$

<br>

**Step 3: 確率遷移行列 $P$ の構築**

講義ノートで学んだ行列 $P = D_c^{-1} X^\top D_r^{-1} X$ を計算します。
これは「アイテム $\to$ 人 $\to$ アイテム」という確率的なつながりを表す行列です。

まず、後半部分 $D_r^{-1} X$（行プロファイル：人の回答確率）を計算します。

$$D_r^{-1} X = \begin{pmatrix} 0.5 & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} 1 & 1 \\ 1 & 0 \end{pmatrix} = \begin{pmatrix} 0.5 & 0.5 \\ 1 & 0 \end{pmatrix}$$

次に、これに $X^\top$ を掛けます（$X^\top (D_r^{-1} X)$）。

$$X^\top = \begin{pmatrix} 1 & 1 \\ 1 & 0 \end{pmatrix} \implies \begin{pmatrix} 1 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 0.5 & 0.5 \\ 1 & 0 \end{pmatrix} = \begin{pmatrix} 1.5 & 0.5 \\ 0.5 & 0.5 \end{pmatrix}$$

最後に、左から $D_c^{-1}$ を掛けて正規化します。

$$P = \begin{pmatrix} 0.5 & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} 1.5 & 0.5 \\ 0.5 & 0.5 \end{pmatrix} = \begin{pmatrix} 0.75 & 0.25 \\ 0.5 & 0.5 \end{pmatrix}$$

**Step 4: 固有値の導出**

行列 $P$ の固有方程式 $|P - \Lambda I| = 0$ を解きます。（固有値を $\Lambda = \lambda^2$ とします）

$$\begin{vmatrix} 0.75 - \Lambda & 0.25 \\ 0.5 & 0.5 - \Lambda \end{vmatrix} = (0.75 - \Lambda)(0.5 - \Lambda) - (0.25 \times 0.5) = 0$$

展開すると：

$$\Lambda^2 - 1.25\Lambda + 0.375 - 0.125 = 0$$

$$\Lambda^2 - 1.25\Lambda + 0.25 = 0$$

因数分解すると：

$$(\Lambda - 1)(\Lambda - 0.25) = 0$$

よって、固有値は $\Lambda = 1, 0.25$ となります。

- $\Lambda_0 = 1$: 自明な解（全てのスコアが定数の場合）。分析では無視します。
- $\Lambda_1 = 0.25$: これが求めたい第1軸の固有値です。

相関係数は $J = \sqrt{\Lambda_1} = \sqrt{0.25} = 0.5$ となります。

**Step 5: 固有ベクトル $\mathbf{b}$ の導出**

$\Lambda_1 = 0.25$ に対応する固有ベクトル（カテゴリスコア） $\mathbf{b} = \begin{pmatrix} b_1 \\ b_2 \end{pmatrix}$ を求めます。
$(P - 0.25 I)\mathbf{b} = \mathbf{0}$ より：

$$\begin{pmatrix} 0.75 - 0.25 & 0.25 \\ 0.5 & 0.5 - 0.25 \end{pmatrix} \begin{pmatrix} b_1 \\ b_2 \end{pmatrix} = \begin{pmatrix} 0.5 & 0.25 \\ 0.5 & 0.25 \end{pmatrix} \begin{pmatrix} b_1 \\ b_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$$

これより $0.5 b_1 + 0.25 b_2 = 0$、すなわち $2b_1 + b_2 = 0 \implies b_2 = -2b_1$ という関係が得られます。

例えば $b_1 = 1$ と置けば、$b_2 = -2$ となります。
つまり、アイテムAのスコアを「1」とすると、アイテムBのスコアは「-2」になる ということです。

これは直感的にも理解できます。

- アイテムA: 両方のユーザーに選ばれている（平均的な存在）。
- アイテムB: ユーザー1にしか選ばれていない（特徴的な存在）。

スコアの符号や大きさが異なることで、これらのアイテムの性質の違いが表現されています。


#### **具体例で確認してみよう！**

以下のPythonコードを実行して、上記の手計算の結果（固有値 1.0, 0.25）が正しいことを確認しましょう。

```python
import numpy as np
import pandas as pd
# ==========================================
# 2人のケース（手計算の確認）
# ==========================================

# 1. データの定義
# User 1: [1, 1] (Item A, Item B)
# User 2: [1, 0] (Item A only)
X = np.array([
    [1, 1],
    [1, 0]
])

# 2. 行列 D_r, D_c の計算
r = X.sum(axis=1) # 行和 [2, 1]
c = X.sum(axis=0) # 列和 [2, 1]

Dr_inv = np.diag(1 / r)
Dc_inv = np.diag(1 / c)

# 3. 確率遷移行列 P の計算
# P = D_c^{-1} * X^T * D_r^{-1} * X
# ステップごとに計算して確認
Row_Profiles = Dr_inv @ X           # 行プロファイル (各ユーザーの確率分布)
Weighted_Sum = X.T @ Row_Profiles   # 列方向への集約
P = Dc_inv @ Weighted_Sum           # 最終的な遷移行列

print("--- 確率遷移行列 P ---")
print(P)

# 4. 固有値・固有ベクトルの計算
eigenvalues, eigenvectors = np.linalg.eig(P)

# 固有値の順序が保証されないため、降順にソート
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print("\n--- 固有値 (Eigenvalues) ---")
print(eigenvalues)

print("\n--- 固有ベクトル (Category Scores b) ---")
print(eigenvectors)

# 第1主成分（2番目の固有値）のベクトルを確認
# b2 = -2 * b1 の関係になっているかチェック
b = eigenvectors[:, 1]  # index 0 は自明解(λ=1)、index 1 が第1軸(λ=0.25)
ratio = b[1] / b[0]
print(f"\n--- 第1軸のスコア比 (b2 / b1) ---")
print(f"{ratio:.2f}  (理論値: -2.00)")
```

数量化Ⅲ類（対応分析）では、固有値問題を解くことで複数の「軸（主成分）」が得られます。 最も相関が高い（情報を多く含む）軸から順に第1軸、第2軸…と定義され、バイプロットでは通常、この上位2つの軸を使ってデータを2次元平面上に配置します。具体的には、あるサンプル $i$ の座標は $(a_i^{(1)}, a_i^{(2)})$、あるカテゴリ $j$ の座標は $(b_j^{(1)}, b_j^{(2)})$ としてプロットされます。これにより、**「人と項目」を共通の基準（軸）** で評価し、その関係性を視覚化しています。サンプル数が3のケースで確認してみましょう。

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# ==========================================
# 3人のケースとバイプロット
# ==========================================

# 1. データの定義 (3人 x 3項目: Chain構造)
# ケース１
# User 1: A, B を選択
# User 2: B, C を選択
# User 3: C    を選択
X = np.array([
    [1, 1, 0],  # User 1
    [0, 1, 1],  # User 2
    [0, 0, 1]   # User 3
])
users = ['User 1', 'User 2', 'User 3']
items = ['Item A', 'Item B', 'Item C']

print("--- データ行列 X ---")
print(X)

# 2. 行列の準備
r = X.sum(axis=1)
c = X.sum(axis=0)
Dr_inv = np.diag(1 / r)
Dc_inv = np.diag(1 / c)

# 3. 固有値問題の解決 (双対性を利用して a と b 両方を求める)

# (A) カテゴリスコア b の計算: P_col = D_c^{-1} X^T D_r^{-1} X
P_col = Dc_inv @ X.T @ Dr_inv @ X
eig_val_b, eig_vec_b = np.linalg.eig(P_col)

# ソート
idx_b = eig_val_b.argsort()[::-1]
eig_val_b = eig_val_b[idx_b]
eig_vec_b = eig_vec_b[:, idx_b]

# (B) サンプルスコア a の計算: P_row = D_r^{-1} X D_c^{-1} X^T
P_row = Dr_inv @ X @ Dc_inv @ X.T
eig_val_a, eig_vec_a = np.linalg.eig(P_row)

# ソート
idx_a = eig_val_a.argsort()[::-1]
eig_val_a = eig_val_a[idx_a]
eig_vec_a = eig_vec_a[:, idx_a]

# 4. スコアの抽出 (第1軸と第2軸)
# index 0 は自明解(λ=1)なので、index 1 (第1軸) と index 2 (第2軸) を使用
# ※ 固有値の大きさ(寄与率)に合わせてプロットするために、固有値の平方根を掛けてスケーリングすることが一般的です

# 固有値 (共通)
lambda_1 = eig_val_b[1]
lambda_2 = eig_val_b[2]

print(f"\n--- 非自明な固有値 ---\n第1軸: {lambda_1:.4f}, 第2軸: {lambda_2:.4f}")

# カテゴリスコア (Item)
b1 = eig_vec_b[:, 1] * np.sqrt(lambda_1) # Scale by sqrt(lambda)
b2 = eig_vec_b[:, 2] * np.sqrt(lambda_2)

# サンプルスコア (User)
a1 = eig_vec_a[:, 1] * np.sqrt(lambda_1)
a2 = eig_vec_a[:, 2] * np.sqrt(lambda_2)

# 5. 可視化 (バイプロット)
plt.figure(figsize=(6, 6))

# 原点のガイドライン
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)

# カテゴリ(Item)のプロット
plt.scatter(b1, b2, c='red', marker='^', s=100, label='Items')
for i, txt in enumerate(items):
    plt.annotate(txt, (b1[i], b2[i]), xytext=(5, 5), textcoords='offset points', color='red', fontsize=12)

# サンプル(User)のプロット
plt.scatter(a1, a2, c='blue', marker='o', s=100, label='Users')
for i, txt in enumerate(users):
    plt.annotate(txt, (a1[i], a2[i]), xytext=(5, 5), textcoords='offset points', color='blue', fontsize=12)

plt.title(f"Quantification Type III Biplot\n(Eigenvalues: {lambda_1:.2f}, {lambda_2:.2f})")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

#### **具体例で確認してみよう！**

以下のコードを実行して、年代別の商品購入データ（クロス集計表）を生成し、それを対応分析によって「年代」と「商品」の関係性がわかるマップ（バイプロット）に変換する様子を確認しましょう。
マップ上で近くにある「年代」と「商品」は、相関が強い（その年代によく買われている）ことを示しています。

```python
# 必要なライブラリのインストール
!pip install prince japanize-matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import prince
import japanize_matplotlib

# 1. サンプルデータの作成
data = {
    'ゲーム':      [50, 40, 10,  5],
    'ファッション': [45, 50, 20, 10],
    '動画サブスク': [40, 45, 30, 10],
    '旅行':        [20, 30, 50, 45],
    'グルメ':   [ 5, 15, 50, 60],
    'ゴルフ': [ 5, 10, 40, 50]
}
# 行（年代）は4つ
df = pd.DataFrame(data, index=['20代', '30代', '40代', '50代'])

print("--- クロス集計表 (頻度データ) ---")
display(df)

# 2. 対応分析 (CA) の実行
# n_components=2: 2次元に圧縮して可視化するため
ca = prince.CA(
    n_components=2,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)
ca = ca.fit(df)

# 3. 結果の確認
# 固有値 (Inertia) と寄与率
print("\n--- 固有値と寄与率 (Eigenvalues & Explained Inertia) ---")
print(ca.eigenvalues_summary)

# 行スコア (年代の座標)
row_coords = ca.row_coordinates(df)
print("\n--- 行スコア: 年代 (Row Coordinates) ---")
print(row_coords)

# 列スコア (商品の座標)
col_coords = ca.column_coordinates(df)
print("\n--- 列スコア: 商品 (Column Coordinates) ---")
print(col_coords)

# 4. バイプロットによる可視化
plt.figure(figsize=(6, 6))

# 行スコア (年代) のプロット
plt.scatter(row_coords.iloc[:, 0], row_coords.iloc[:, 1], label='年代 (Age Groups)', color='blue', marker='o', s=100)
for i, txt in enumerate(row_coords.index):
    plt.annotate(txt, (row_coords.iloc[i, 0], row_coords.iloc[i, 1]), xytext=(5, 5), textcoords='offset points', color='blue')

# 列スコア (商品カテゴリ) のプロット
plt.scatter(col_coords.iloc[:, 0], col_coords.iloc[:, 1], label='商品 (Product Categories)', color='red', marker='^', s=100)
for i, txt in enumerate(col_coords.index):
    plt.annotate(txt, (col_coords.iloc[i, 0], col_coords.iloc[i, 1]), xytext=(5, 5), textcoords='offset points', color='red')

plt.title("対応分析 (Correspondence Analysis): 年代 vs 商品カテゴリ")
plt.xlabel(f"第1成分 (Component 0)")
plt.ylabel(f"第2成分 (Component 1)")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
