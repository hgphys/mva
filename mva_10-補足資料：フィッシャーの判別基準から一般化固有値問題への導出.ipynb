## **フィッシャーの判別基準から一般化固有値問題への導出**

フィッシャー比 $J(\mathbf{w})$ を最大化する際、なぜ固有値問題が現れるのかを詳細に解説します。

### **1. 目的関数の整理**

最大化したい対象は、レイリー商と呼ばれる以下の形式です。

$$J(\mathbf{w}) = \frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_W \mathbf{w}}$$

ここで、$S_B$ はグループ間変動行列、$S_W$ はグループ内変動行列であり、ともに正方行列かつ対称行列です。

<br>

### **2. 微分による最適化**

スカラー値 $J(\mathbf{w})$ を最大にする $\mathbf{w}$ を見つけるため、ベクトル $\mathbf{w}$ で微分し、その値を $0$ と置きます。

商の微分公式 $\left(\frac{u}{v}\right)' = \frac{u'v - uv'}{v^2}$ を適用すると、極値を持つ条件は分子が $0$ になること、すなわち $u'v = uv'$ です。

$$\frac{\partial (\mathbf{w}^\top S_B \mathbf{w})}{\partial \mathbf{w}} (\mathbf{w}^\top S_W \mathbf{w}) = (\mathbf{w}^\top S_B \mathbf{w}) \frac{\partial (\mathbf{w}^\top S_W \mathbf{w})}{\partial \mathbf{w}}$$

<br>

### **3. ベクトル微分の実行**

二次形式のベクトル微分公式 $\frac{\partial (\mathbf{x}^\top A \mathbf{x})}{\partial \mathbf{x}} = 2A\mathbf{x}$ を適用します。

$$(2 S_B \mathbf{w}) (\mathbf{w}^\top S_W \mathbf{w}) = (\mathbf{w}^\top S_B \mathbf{w}) (2 S_W \mathbf{w})$$

<br>

### **4. 固有値問題への変形**

両辺を $2 (\mathbf{w}^\top S_W \mathbf{w})$ で割ります。

$$S_B \mathbf{w} = \frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_W \mathbf{w}} S_W \mathbf{w}$$

ここで、係数部分 $\frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_W \mathbf{w}}$ は、最大化したい指標 $J(\mathbf{w})$ そのものです。これを固有値 $\lambda$ と置くと：

$$S_B \mathbf{w} = \lambda S_W \mathbf{w}$$

これは 一般化固有値問題 の定義そのものです。さらに左から $S_W^{-1}$ を掛けることで（$S_W$ が正則であると仮定）：

$$S_W^{-1} S_B \mathbf{w} = \lambda \mathbf{w}$$

となり、行列 $S_W^{-1} S_B$ の固有値問題を解くことで、目的関数を最大化するベクトル $\mathbf{w}$ が得られることが証明されました。

### **5. なぜ「最大固有値」が「最大値」を与えるのか？**

微分を $0$ と置いただけでは、最小値や鞍点も含まれてしまいます。しかし、得られた一般化固有値問題の式 $S_B \mathbf{w} = \lambda S_W \mathbf{w}$ に対して、左から $\mathbf{w}^\top$ を掛けてみましょう。

$$\mathbf{w}^\top S_B \mathbf{w} = \lambda \mathbf{w}^\top S_W \mathbf{w}$$

これを $\lambda$ について解くと：

$$\lambda = \frac{\mathbf{w}^\top S_B \mathbf{w}}{\mathbf{w}^\top S_W \mathbf{w}} = J(\mathbf{w})$$

つまり、固有値 $\lambda$ の値そのものが、その時のフィッシャー比 $J(\mathbf{w})$ の値になっています。

したがって、

- 最大固有値 $\lambda_{\max}$ に対応する固有ベクトルを採用すれば、それが 最大化 の解となる。
- 最小固有値 $\lambda_{\min}$ に対応する固有ベクトルを採用すれば、それが 最小化 の解となる。

この性質があるため、判別分析では「一般化固有値問題の最大固有値」を求めることが、フィッシャー比を最大化することと等価になるのです。
